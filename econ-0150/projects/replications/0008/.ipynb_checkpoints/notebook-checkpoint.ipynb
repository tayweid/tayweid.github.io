**Research Question:** What is the relationship between Voter Turnout and the Margin of U.S. Presidential Victories by County?

**Data Source:** County presidential election results (2000-2024) merged with county population estimates via FIPS codes

**Methods:** OLS regression with turnout as predictor of vote margin

**Main Finding:** The analysis examines whether higher voter turnout is associated with more Democratic or Republican vote margins across thousands of U.S. counties from 2000 to 2024.

**Course Concepts Used:**
- Merging datasets on FIPS codes
- Creating new variables (turnout percentage, margin percentage)
- OLS regression
- Residual analysis

---
## Step 0 | Setup

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf

# File Path
file_path = 'https://tayweid.github.io/econ-0150/projects/replications/0008/data/'
```

```python
pres_data = pd.read_csv(file_path + 'countypres_2000-2024.csv')
pres_data.head()
```

```python
pop_data = pd.read_csv(file_path + 'Population%20Estimates%20-%20US,%20States,%20Counties.csv')
pop_data.head()
```

---
## Step 1 | Data Preparation

```python
# Filter to major parties and clean FIPS codes
major_parties = pres_data[pres_data['party'].isin(['DEMOCRAT', 'REPUBLICAN'])].copy()
major_parties = major_parties.dropna(subset=['county_fips'])
major_parties['county_fips'] = major_parties['county_fips'].astype(int)

# Pivot to get Democrat and Republican vote columns per county-year
party_votes = major_parties.groupby(
    ['year', 'county_fips', 'state_po', 'county_name', 'party']
)['candidatevotes'].sum().reset_index()

pivoted = party_votes.pivot_table(
    index=['year', 'county_fips', 'state_po', 'county_name'], 
    columns='party', 
    values='candidatevotes'
).reset_index()

pivoted = pivoted.fillna(0)
pivoted.head()
```

```python
# Get total votes per county-year
total_votes = pres_data.dropna(subset=['county_fips']).copy()
total_votes['county_fips'] = total_votes['county_fips'].astype(int)
total_votes = total_votes.groupby(['year', 'county_fips'])['totalvotes'].max().reset_index()

# Merge with pivoted data
election_data = pd.merge(pivoted, total_votes, on=['year', 'county_fips'])

print(f"Election data: {len(election_data)} county-year observations")
election_data.head()
```

```python
# Prepare population data using FIPS codes for reliable matching
# Filter to county-level estimates (exclude state and national totals)
pop_counties = pop_data[
    (pop_data['IBRC_Geo_ID'] % 1000 != 0) &
    (pop_data['Count or Estimate'] == 'Estimate') &
    (pop_data['State or County Release'] == 'County')
].copy()

# Average population per county-year (handle duplicate estimates)
pop_agg = pop_counties.groupby(['IBRC_Geo_ID', 'Year'])['Population'].mean().reset_index()
pop_agg.columns = ['county_fips', 'year', 'population']

print(f"Population estimates: {len(pop_agg)} county-year observations")
pop_agg.head()
```

```python
# Merge election and population data on FIPS code and year
data = pd.merge(election_data, pop_agg, on=['county_fips', 'year'], how='inner')

# Calculate turnout percentage
data['turnout_pct'] = (data['totalvotes'] / data['population']) * 100

# Calculate vote margin percentage (positive = Democratic lead)
data['margin_pct'] = ((data['DEMOCRAT'] - data['REPUBLICAN']) / data['totalvotes']) * 100

# Filter to reasonable turnout values
data = data[(data['turnout_pct'] > 5) & (data['turnout_pct'] <= 100)]

print(f"Final dataset: {len(data)} county-year observations")
data.head()
```

---
## Step 2 | Data Exploration

```python
# Summary statistics
print("Summary Statistics:")
print(data[['turnout_pct', 'margin_pct']].describe())
```

```python
# Average turnout and margin by election year
print("Averages by election year:")
print(data.groupby('year')[['turnout_pct', 'margin_pct']].mean().round(2))
```

---
## Step 3 | Visualization

```python
# Scatter plot: Turnout vs Vote Margin
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='turnout_pct', y='margin_pct', 
                hue='year', palette='viridis', alpha=0.2, s=5)
sns.regplot(data=data, x='turnout_pct', y='margin_pct', 
            scatter=False, color='blue')

plt.axhline(0, color='red', linestyle='--', linewidth=0.8)
plt.xlabel('Voter Turnout (%)')
plt.ylabel('Democratic Margin (D - R) %')
plt.title('Voter Turnout vs. Presidential Vote Margin by County')
plt.legend(title='Election Year', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

---
## Step 4 | Statistical Analysis

```python
# OLS regression: Vote margin ~ Turnout
model = smf.ols('margin_pct ~ turnout_pct', data=data).fit()
print("OLS Regression: Democratic Margin ~ Voter Turnout")
print(model.summary())
```

```python
# Residual plot
data['predicted'] = model.predict()
data['residuals'] = data['margin_pct'] - data['predicted']

plt.figure(figsize=(10, 6))
plt.scatter(data['predicted'], data['residuals'], alpha=0.1, s=5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Democratic Margin (%)')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.tight_layout()
plt.show()
```

```python
# Histogram of residuals
plt.figure(figsize=(10, 6))
sns.histplot(data['residuals'], bins=50, kde=True)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.show()
```

---
## Step 5 | Results Interpretation

### Key Findings

The regression tests whether voter turnout predicts the Democratic-Republican vote margin at the county level across seven presidential elections (2000-2024).

- A **positive** coefficient on turnout means higher turnout is associated with more Democratic margins
- A **negative** coefficient means higher turnout is associated with more Republican margins
- The R-squared indicates how much of the variation in vote margins is explained by turnout alone

### Economic Context

- **Voter mobilization:** Campaigns invest heavily in turnout operations, especially in competitive counties
- **Composition effects:** Higher turnout may bring different types of voters to the polls
- **County heterogeneity:** Urban counties tend to have both higher turnout infrastructure and different partisan leanings

### Caveats

- Correlation does not imply causation  many confounding factors drive both turnout and partisan lean
- County-level analysis masks within-county variation
- Turnout is calculated using total population, not voting-eligible population, which understates true turnout rates
- The relationship may differ across urban/rural counties and across election years

## Step 0 | Imports and Setup

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
# Open a dataset from your project folder in Google Drive
data = pd.read_csv('econ_final_data.csv')
data
```

## Step 1 | Data Preparation

Turn your data into something you can explore through data operations.

```python
# Converted GDP Per Capita into Thousands of Dollars in GDP Per Capita, then round to one decimal place, i.e to the hundreds. This makes the data cleaner/nicer to work with.
data['K_GDP'] = round(data['GDP_Per_Cap'] / 1000, 1)
data
```

```python
# Save your dataset after you've worked on it
data.to_csv('cleaned_data.csv')
```

## Step 2 | Exploratory Data Analysis

Summarize the main variables and relationships of interest.

```python
# Exploratory Data Analysis here
sns.boxplot(data = data, x = 'Republican', y = 'Birth_Rate', hue='Republican', palette={0: 'blue', 1: 'red'}, legend=False)
plt.xlabel('Republican')
plt.ylabel('Teen Birth Rate')
# Simple side by side box plots to show only the relationship between state political leaning and teen birth rate.
```

## Step 3 (Part 3 - 5)

Develop your statistical model.

1. Write down you null hypothesis
2. Write down your statistical model
3. Run your statistical model
4. Visualize your statistical model (if appropriate)

```python
interactionmodel = smf.ols('Birth_Rate ~ K_GDP + Republican + K_GDP:Republican', data = data).fit()
interactionmodel.summary()
# Testing out a model with an interaction term, to see if teen birth rate changes with GDP per capita at different rates for republican and democrat states.
# At alphas of .05, neither the coefficent for K_GDP nor for K_GDP*Republican was significant, with p-values of .498 and .059, respectfully.
# This means a model that includes the interaction term is not a good fit, and the states share the same slope no matter their politcal leaning.
```

```python
model = smf.ols('Birth_Rate ~ K_GDP + Republican', data = data).fit()

model.summary()
# Null hypothesis that B_0 = 0, rejected because p value is 0.000, so B_1 is non-zero
# Null hypothesis that B_1 = 0, rejected because p value is 0.012, so B_1 is non-zero
# Null hypothesis that B_2 = 0, rejected because p value is 0.001, so B_2 is non-zero
```

```python
# Model visualization
sns.scatterplot(data = data, x = 'K_GDP', y = 'Birth_Rate', hue='Republican', palette={0: 'blue', 1: 'red'}, legend=False)
# Scatterplot, points are color coded by politcal leaning.
plt.xlabel('GDP Per Capita (Thousands of Dollars)')
plt.ylabel('Teen Birth Rate')
plt.plot([0, 100 ], [19.3933+4.6480 , 19.3933+4.6480 - (100*0.1431)], color = 'red')
plt.plot([0, 100 ], [19.3933 , 19.3933 - (100*0.1431)], color = 'blue')
# Creates parallel regression line, top red one is for Republican states, bottom blue one is for Democratic states.
plt.xlim(30,100)
plt.ylim(0, 30)
```

```python
gdp_resid = sns.residplot(x=data['K_GDP'], y=model.resid)
# Check residuals
```

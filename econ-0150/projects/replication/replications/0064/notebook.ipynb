```python
import statsmodels.api as sm
import numpy as np

# Your data
x = np.array(attendance)
y = np.array(winning_pct)

# Add constant term (intercept)
x_with_const = sm.add_constant(x)

# Fit regression
model = sm.OLS(y, x_with_const).fit()

# Extract statistics
slope_t_stat = model.tvalues[1]     # t-statistic for slope
slope_p_value = model.pvalues[1]    # p-value for slope
f_stat = model.fvalue                # F-statistic
f_p_value = model.f_pvalue           # p-value for F-test

print("T-statistic (slope):", slope_t_stat)
print("P-value (slope):", slope_p_value)
print("F-statistic:", f_stat)
print("F-test p-value:", f_p_value)

# Optional: full regression summary (like R or Stata)
print(model.summary())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Setup your data
# (Replace these with your actual columns, e.g., df['Attendance'])
attendance = np.array([80806, 53479, 59931, 47686, 56916, 26183, 38999, 65632, 38216, 53077, 50863, 39130, 39839, 29737, 39173, 32652, 26035]).reshape(-1, 1)
winning_pct = np.array([0.714, 0.167, 0.769, 0.462, 0.462, 0.692, 0.417, 0.462, 0.538, 0.538, 0.692, 0.769, 0.538, 0.333, 0.462, 0.786, 0.250])

# 2. Create and fit the model
model = LinearRegression()
model.fit(attendance, winning_pct)

# 3. Calculate Predictions
predictions = model.predict(attendance)

# 4. CALCULATE RESIDUALS (The core step)
# Residual = Actual Value - Predicted Value
residuals = winning_pct - predictions

# 5. Plot the Residuals
plt.figure(figsize=(10, 6))

# The "Zero Line" represents the regression line.
# Points above it are positive residuals, points below are negative.
plt.axhline(y=0, color='black', linestyle='--')

# Plot the residuals as dots
plt.scatter(attendance, residuals, color='blue', s=100)

# Optional: Draw vertical lines to show the distance from 0
plt.vlines(attendance, 0, residuals, colors='red', linestyle='dotted')

plt.title('Residual Plot')
plt.xlabel('Average Attendance')
plt.ylabel('Residuals (Difference from Model)')
plt.show()
```

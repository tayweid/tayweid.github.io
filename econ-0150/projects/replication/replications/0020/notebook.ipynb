## Step 3 (Part 3 - 5)

Develop your statistical model.

Null Hypothesis: There is no relationship between major and income as age increases

Statistical model: B_0 + B_1DEG + B_2AGE + B_3EDUC + e

```python
model = smf.ols("logINCTOT ~ C(DEGFIELD) + AGE + EDUC", data=data).fit()
print(model.summary())
```

```python
data["predicted_logINCTOT"] = model.predict(data)
data["predicted_INCTOT"] = np.exp(data["predicted_logINCTOT"])

ax = sns.boxplot(data=data, x="DEGFIELD_LABEL", y="predicted_logINCTOT", showfliers=False)
plt.xticks(rotation=90)

# Highlight median lines in red
# For each box, there are 5 lines: 2 whiskers, 2 caps, 1 median.
# The median is typically the 5th line for each box (index 4, 9, 14, ...)
for i, median_line in enumerate(ax.lines):
    if i % 5 == 4: # Select every 5th line, starting from the 4th (0-indexed)
        median_line.set_color('red')
        median_line.set_linewidth(2)

plt.xlabel('Major')
plt.ylabel('Predicted Income')
plt.title('Predicted Income Range According to Major')
plt.show()
```

## Step 0 | Imports and Setup

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv('Merged_GDP_Medals.csv')
data.head(87)
```

```python
sns.scatterplot(data=data, x='Differences_medal', y='Differences_cap')
plt.xlabel('Differences in Medal count')
plt.ylabel('Differences in GDP per Cap')
plt.title('Differences in Medal count vs Differences in GDP per Cap')
plt.show()




```

```python
#make a histogram of the differences in gdp per cap
sns.histplot(data=data, x='Differences_cap')
plt.xlabel('Differences in GDP per Cap')
plt.ylabel('Frequency')
plt.title('Histogram of Differences in GDP per Cap')
plt.show()
```

```python
#create a residual plot
sns.residplot(data=data, x='Differences_medal', y='Differences_cap')
```

```python
#create a interecept table with this data so we can see the p-value and the interecpt
model = smf.ols(formula='Differences_medal ~ Differences_cap', data=data)
results = model.fit()
print(results.summary())
```

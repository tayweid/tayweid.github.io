```python
import pandas as pd

df = pd.read_csv("international-tourist-trips.csv")

# View the data
df.head()
```

```python
brazil = df[df['Entity'] == 'Brazil']
brazil = brazil[['Year', 'Inbound arrivals of tourists']].dropna()

brazil.head()
```

```python
import statsmodels.api as sm

# Define variables
X = brazil['Year']
y = brazil['Inbound arrivals of tourists']

# Add intercept
X = sm.add_constant(X)

# Fit GLM
glm_model = sm.GLM(y, X, family=sm.families.Gaussian())
result = glm_model.fit()
print(result.summary().tables[1])
```

```python
import matplotlib.pyplot as plt

plt.scatter(brazil['Year'], y)
plt.axvline(x=2016, linestyle='--', color='red')
plt.xlabel("Year")
plt.ylabel("Inbound Tourist Arrivals")
plt.title("GLM  Brazil Tourism vs Year")
plt.show()
```

Combine Plots

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

countries_data = {
    'Brazil': {'df': brazil, 'olympic_years': [2016], 'color': 'red'},
    'USA': {'df': USA, 'olympic_years': [1996, 2002], 'color': 'blue'},
    'China': {'df': China, 'olympic_years': [2008], 'color': 'green'},
    'UK': {'df': UK, 'olympic_years': [2012], 'color': 'purple'},
    'Canada': {'df': Canada, 'olympic_years': [2010], 'color': 'orange'}
}

for country_name, data in countries_data.items():
    df = data['df']
    color = data['color']
    olympic_years = data['olympic_years']

    plt.scatter(df['Year'], df['Inbound arrivals of tourists'], label=f'{country_name} Tourist Arrivals', color=color)

    for year in olympic_years:
        plt.axvline(x=year, linestyle='--', color=color, label=f'{country_name} Olympics ({year})')

plt.xlabel("Year")
plt.ylabel("Inbound Tourist Arrivals")
plt.title("Inbound Tourist Arrivals vs. Year with Olympic Events")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

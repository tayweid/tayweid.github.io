```python
# prompt: Location A: Wait times follow a normal distribution with mean (μ) = 12 minutes and standard deviation (σ) = 2.5 minutes

import scipy.stats as stats

# Probability of waiting less than 10 minutes at Location A
prob_less_than_10 = stats.norm.cdf(10, loc=12, scale=2.5)
print(f"Probability of waiting less than 10 minutes at Location A: {prob_less_than_10}")

# Probability of waiting more than 15 minutes at Location A
prob_more_than_15 = 1 - stats.norm.cdf(15, loc=12, scale=2.5)
print(f"Probability of waiting more than 15 minutes at Location A: {prob_more_than_15}")

# Probability of waiting between 10 and 15 minutes at Location A
prob_between_10_and_15 = stats.norm.cdf(15, loc=12, scale=2.5) - stats.norm.cdf(10, loc=12, scale=2.5)
print(f"Probability of waiting between 10 and 15 minutes at Location A: {prob_between_10_and_15}")
```

```python
# prompt: a) Create histograms representing the wait time distributions from both locations on the same figure. Add vertical lines showing the mean wait time for each location. Use appropriate colors and labels to make the comparison clear.

import matplotlib.pyplot as plt
import numpy as np

# Location A data (from previous code)
mean_a = 12
std_a = 2.5
wait_times_a = np.random.normal(loc=mean_a, scale=std_a, size=10000)

# Location B data (example data, replace with actual data if available)
mean_b = 15  # Example mean
std_b = 3   # Example standard deviation
wait_times_b = np.random.normal(loc=mean_b, scale=std_b, size=10000)


# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(wait_times_a, bins=100, alpha=0.5, color='blue', label='Location A')
plt.hist(wait_times_b, bins=100, alpha=0.5, color='red', label='Location B')

# Add vertical lines for the means
plt.axvline(mean_a, color='blue', linestyle='dashed', linewidth=2, label=f'Mean A: {mean_a:.2f}')
plt.axvline(mean_b, color='red', linestyle='dashed', linewidth=2, label=f'Mean B: {mean_b:.2f}')

# Add labels and title
plt.xlabel('Wait Time (minutes)')
plt.ylabel('Frequency')
plt.title('Wait Time Distributions at Two Locations')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: For Location A, simulate drawing 100 random samples of size 30 and calculate the mean of each sample. Create a histogram of these sample means.

import matplotlib.pyplot as plt
import numpy as np

# Location A parameters
mean_a = 12
std_a = 2.5
sample_size = 30
num_samples = 100

# Simulate drawing samples and calculate sample means
sample_means = []
for _ in range(num_samples):
  sample = np.random.normal(loc=mean_a, scale=std_a, size=sample_size)
  sample_means.append(np.mean(sample))

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=20, edgecolor='black')  # Adjust bins as needed
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Histogram of Sample Means (Location A)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot A: Clear heteroskedasticity (increasing variance)
fitted_values_a = np.linspace(2, 10, 50)
residuals_a = np.random.normal(0, 0.2 + 0.3 * fitted_values_a/10, 50)
# Make the pattern more obvious
for i in range(len(fitted_values_a)):
    residuals_a[i] *= (1 + fitted_values_a[i] * 0.1)

ax1.scatter(fitted_values_a, residuals_a, alpha=0.7, color='blue')
ax1.axhline(y=0, color='r', linestyle='-')
ax1.set_xlabel('Fitted Values', fontsize=12)
ax1.set_ylabel('Residuals', fontsize=12)
ax1.set_title('Plot A: Residuals vs. Fitted Values', fontsize=14)

# Plot B: Clear non-linearity (U-shaped pattern)
fitted_values_b = np.linspace(1, 9, 50)
# Create U-shaped pattern
residuals_b = 0.8 * (fitted_values_b - 5)**2 / 8 - 0.4 + np.random.normal(0, 0.4, 50)

ax2.scatter(fitted_values_b, residuals_b, alpha=0.7, color='blue')
ax2.axhline(y=0, color='r', linestyle='-')
ax2.set_xlabel('Fitted Values', fontsize=12)
ax2.set_title('Plot B: Residuals vs. Fitted Values', fontsize=14)

plt.tight_layout()
plt.savefig('ME_07_demo_residual_plots.png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot A: Non-linear relationship (U-shaped residuals)
x_a = np.linspace(0, 10, 50)
predictions_a = np.linspace(1, 6, 50)
residuals_a = 0.5 * (x_a - 5)**2 + np.random.normal(0, 0.5, 50)

ax1.scatter(predictions_a, residuals_a, alpha=0.7)
ax1.axhline(y=0, color='r', linestyle='-')
ax1.set_xlabel('Fitted Values ($\hat{y}$)')
ax1.set_ylabel('Residuals ($\hat{\\varepsilon}$)')
ax1.set_title('Plot A')

# Plot B: Heteroskedastic pattern (fan shape)
x_b = np.linspace(0, 10, 50)
predictions_b = np.linspace(1, 6, 50)
residuals_b = np.random.normal(0, 0.2 + 0.3 * x_b/10, 50)

ax2.scatter(predictions_b, residuals_b, alpha=0.7)
ax2.axhline(y=0, color='r', linestyle='-')
ax2.set_xlabel('Fitted Values ($\hat{y}$)')
ax2.set_title('Plot B')

plt.tight_layout()
plt.savefig('ME_07_residual_plot.png', dpi=300)
plt.show()
```

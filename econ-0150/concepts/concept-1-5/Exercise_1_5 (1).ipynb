## Exercise 1.5 | Filtering

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/concepts/concept-1-5/data/'
```

```python
# Load Dataset
hours = pd.read_csv(file_path + 'Starbucks_Location_Hours.csv', index_col=0)

# Figure
plt.figure(figsize=(10, 4))
sns.histplot(data=hours, x='open', bins=range(0,25,1), edgecolor='black')

# Styling
plt.xlabel('Hour of the Day')
plt.xlim(0,24)
plt.ylabel('')
plt.yticks([])
plt.title('Opening Times')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.tight_layout()
```

```python
# Load Dataset
hours = pd.read_csv(file_path + 'Starbucks_Location_Hours.csv', index_col=0)
hours = hours[hours.COUNTRY_CODE == 'US']

# Figure
plt.figure(figsize=(10, 4))
sns.histplot(data=hours, x='open', bins=range(0,25,1), edgecolor='black')

# Styling
plt.xlabel('Hour of the Day')
plt.xlim(0,24)
plt.ylabel('')
plt.yticks([])
plt.title('Opening Times (US)')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.tight_layout()
```

## Combining Filters

```python
# Load Dataset
hours = pd.read_csv(file_path + 'Starbucks_Location_Hours.csv', index_col=0)
hours = hours[hours.COUNTRY_CODE.isin(['CA','US'])]

# Figure
plt.figure(figsize=(10, 4))
sns.histplot(data=hours, x='open', bins=range(0,25,1), edgecolor='black')

# Styling
plt.xlabel('Hour of the Day')
plt.xlim(0,24)
plt.ylabel('')
plt.yticks([])
plt.title('Opening Times (in US or Canada)')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.tight_layout()

# Save Figure
#plt.savefig('i_03.png')
```

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-1-6/data/'
```

```python
geo = pd.read_csv(file_path + 'US_Cities.csv', index_col=0)
geo.head()
```

```python
sns.scatterplot(geo, x='lng', y='lat')
```

```python
sns.scatterplot(geo, x='lng', y='lat', hue='timezone')
```

```python
sns.scatterplot(
    geo, x='lng', y='lat',
    size='population', sizes=(1, 200),
    hue='avg_temp', legend=False
)
```

```python
east = geo[geo.timezone == 'America/New_York']
sns.scatterplot(east, x='lat', y='avg_temp', size='p', sizes=(0,100))
```

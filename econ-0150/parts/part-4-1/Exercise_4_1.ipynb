# Exercise 4.1 | Numerical Predictors

Are wealthier countries happier? We'll use the World Happiness Report data to fit a bivariate GLM with a numerical predictor:

$$y_i = \beta_0 + \beta_1 x_i + \epsilon_i$$

Our outcome ($y$) is **Life Evaluation** (0–10 scale) and our predictor ($x$) is **log10 GDP per capita**.

```python
# Imports
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-4-1/data/'
```

We're exploring the dataset `Life_Evaluation_GDP_PerCap.csv` from the World Happiness Report collected by Gallup World Poll and GDP Per capita from the World Bank.

```python
# Load Data
data = pd.read_csv(file_path + 'Life_Evaluation_GDP_PerCap.csv')
# Log10
data['log_GDP_PerCap'] = np.log10(data['GDP_Per_Cap'])
# Filter for 2024
data = data[data['Year'] == 2024]
# Display header
data.head()
```

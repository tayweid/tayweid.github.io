# Exercise 4.1 | Bivariate GLM with Numerical Predictor

We will test a GLM with an intercept and a slope.

$$y_i = \beta_0 + \beta_1 x_i$$

```python
# Imports
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-4-1/data/'
```

## Exploratory Data Analysis

```python
# Load
data = pd.read_csv(file_path + 'Life_Evaluation_GDP_PerCap.csv')
# Log Transformation
data['log_GDP_PerCap'] = np.log10(data['GDP_Per_Cap'])
# Filter: 2024
data = data[data['Year'] == 2024]
# Show
data.head()
```

```python
# Visualize: Numerical x Numerical
sns.scatterplot(y='Life_Evaluation', x='log_GDP_PerCap', data=data)
```

## General Linear Model

```python
# Model: y = b + mx
model = smf.ols('Life_Evaluation ~ log_GDP_PerCap', data).fit() # Intercept is included by default
print(model.summary().tables[1])
```

```python
# Visualize: Numerical x Numerical
sns.regplot(data=data, y='Life_Evaluation', x='log_GDP_PerCap', ci=None, line_kws={'color': 'red'})
```

## Predictions

```python
# Parameters
b0, b1 = model.params  # intercept, slope
```

```python
# Predict: Log GDP of 3
prediction = b0 + b1 * 3
print(prediction)
```

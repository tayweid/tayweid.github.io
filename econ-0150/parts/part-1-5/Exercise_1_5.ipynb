# Exercise 1.5 | Bivariate Relationships


```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
file_path = 'https://tayweid.github.io/econ-0150/parts/part-1-5/data/'
```

## Coffee Production and GDP

```python
gdp = pd.read_csv(file_path + 'Beans_GDP_2019.csv')
```

```python
gdp
```

```python
sns.scatterplot(gdp, x='GDP', y='coffee_prod')
plt.title('Coffee Production and GDP in 2019')
plt.grid()
```

## Coffee and Oil Prices

```python
prices = pd.read_csv(file_path + 'Coffee_Oil.csv')
prices.head()
```

```python
sns.scatterplot(prices, x='Coffee', y='Oil')
plt.title('Coffee and Oil Prices')
plt.grid()
```

## Q2. Boxplots and Quartiles

Use the class survey dataset to create a boxplot with stripplot of "Approximately how many miles away from Pittsburgh is your hometown?"

**a) Create a boxplot with stripplot.**

```python
sns.boxplot(survey, x='Approximately how many miles away from Pittsburgh is your hometown?', whis=(0, 100))
sns.stripplot(survey, x='Approximately how many miles away from Pittsburgh is your hometown?')
plt.title('Distance from Pittsburgh')
```

**b) What is the median distance from Pittsburgh?**

```python
survey['Approximately how many miles away from Pittsburgh is your hometown?'].median()
```

**c) Explain why the mean (from Q1) and the median differ. (Hint: Think about the shape of the distribution.)**

The distribution is right-skewed â€” most students are relatively close to Pittsburgh, but a few students are very far away (e.g., international students). These outliers pull the mean up but don't affect the median, which only depends on the middle position. This is why the mean is much larger than the median.

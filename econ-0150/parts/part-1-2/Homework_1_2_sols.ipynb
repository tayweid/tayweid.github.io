## Q2. Summarizing 'When is your birthyear?'

Use the class survey dataset to answer the following questions.

a) Appropriately visualize the variable 'When is your birthyear?'.

```python
# Visualize 'When is your birthyear?' as a histogram
sns.histplot(data, x='When is your birthyear?', bins=range(2000,2012,1))

# Optional Styling
plt.xlabel('')
plt.yticks([])
plt.ylabel('')
plt.title('When is your birthyear?')
sns.despine(left=True, right=True, trim=True)
```

## Q3. Summarizing 'How excited are you for this class?'

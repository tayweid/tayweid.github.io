## Q1. Histograms: Choosing Bin Sizes

Use the class survey dataset to create a histogram of "Approximately how many miles away from Pittsburgh is your hometown?"

**a) Create a histogram with bins of width 100 miles. Save this figure.**

```python
# Note: Column name may vary - adjust as needed based on actual survey data
sns.histplot(survey, x='Approximately how many miles away from Pittsburgh is your hometown?', bins=range(0, 10000, 100))
plt.title('Distance from Pittsburgh (100-mile bins)')
plt.savefig('hw_1_2_q1a.png')
```

**b) Create a histogram with bins of width 500 miles. Save this figure.**

```python
sns.histplot(survey, x='Approximately how many miles away from Pittsburgh is your hometown?', bins=range(0, 10000, 500))
plt.title('Distance from Pittsburgh (500-mile bins)')
plt.savefig('hw_1_2_q1b.png')
```

**c) Which bin width better shows the distribution? Explain your reasoning in one sentence.**

**Answer:** The 500-mile bins likely work better because with a small class size, 100-mile bins may be too narrow and introduce noise, while 500-mile bins provide a clearer picture of the overall distribution.

## Q2. Describing Center and Spread

Use the class survey dataset to answer the following questions about "Approximately how many miles away from Pittsburgh is your hometown?"

**a) Calculate the mean distance from Pittsburgh.**

```python
mean_distance = survey['Approximately how many miles away from Pittsburgh is your hometown?'].mean()
print(f"Mean distance: {mean_distance:.1f} miles")
```

**b) Calculate the standard deviation of distance from Pittsburgh.**

```python
sd_distance = survey['Approximately how many miles away from Pittsburgh is your hometown?'].std()
print(f"Standard deviation: {sd_distance:.1f} miles")
```

**c) Write a one-sentence interpretation using both values.**

(Example format: "The average student is about X miles from Pittsburgh; distances typically vary by about Y miles from that average.")

**Answer:** The average student is about [mean] miles from Pittsburgh; distances typically vary by about [SD] miles from that average. (Values depend on actual data)

## Q3. Boxplots and Quartiles

Use the class survey dataset to create a boxplot with stripplot of "Approximately how many miles away from Pittsburgh is your hometown?"

**a) Create a boxplot with stripplot. Save this figure.**

```python
sns.boxplot(survey, x='Approximately how many miles away from Pittsburgh is your hometown?', whis=(0,100))
sns.stripplot(survey, x='Approximately how many miles away from Pittsburgh is your hometown?')
plt.title('Distance from Pittsburgh')
plt.savefig('hw_1_2_q3a.png')
```

**b) What is the median distance from Pittsburgh?**

```python
median_distance = survey['Approximately how many miles away from Pittsburgh is your hometown?'].median()
print(f"Median distance: {median_distance:.1f} miles")
```

## Q4. Comparing Summary Approaches

**a) Compare the mean and median distances from Q2 and Q3. Which is larger?**

**Answer:** The mean is likely larger than the median because of students from far away (e.g., international students) pulling the mean up.

**b) Based on the histogram and boxplot you created, explain why the mean and median differ. (Hint: Think about the shape of the distribution.)**

**Answer:** The distribution is right-skewed — most students are relatively close to Pittsburgh, but a few students are very far away (outliers). These outliers pull the mean up but don't affect the median, which only depends on the middle position.

## Q5. Choosing the Right Tool

Use the class survey dataset to visualize "When is your birthyear?"

**a) Create a histogram of birthyear. Save this figure.**

```python
sns.histplot(survey, x='When is your birthyear?')
plt.title('Student Birthyears')
plt.savefig('hw_1_2_q5a.png')
```

**b) Create a boxplot with stripplot of birthyear. Save this figure.**

```python
sns.boxplot(survey, x='When is your birthyear?', whis=(0,100))
sns.stripplot(survey, x='When is your birthyear?')
plt.title('Student Birthyears')
plt.savefig('hw_1_2_q5b.png')
```

**c) Given the number of students in our class, which visualization better shows the distribution of birthyears? Explain your reasoning in one sentence.**

**Answer:** The boxplot + stripplot works better because with a small class size (~30-50 students), the histogram bins become noisy or sparse, while the stripplot lets us see each individual student and the boxplot shows the quartiles clearly.

**d) If we had 5,000 students instead, would your answer change? Why or why not?**

**Answer:** Yes, with 5,000 students a histogram would work better. With many observations, histogram bins would have enough data points to show the shape of the distribution smoothly, while the stripplot would become overcrowded and hard to read.

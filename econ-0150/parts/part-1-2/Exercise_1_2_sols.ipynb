# Exercise 1.2 | Numerical Variables

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-1-2/data/'
```

## Histograms: $40k Income Customers

```python
customers = pd.read_csv(file_path + 'Starbucks_Customer_Profiles_40k.csv', index_col=0)
```

```python
plt.figure(figsize=(10, 4), dpi=400)
sns.histplot(customers, x='age')
plt.savefig('i/ex_01.png')
```

## Boxplot + Stripplot: Coffee Per Capita

```python
coffee = pd.read_csv(file_path + 'Coffee_Per_Cap_2019.csv', index_col=0)
```

```python
plt.figure(figsize=(10, 4), dpi=400)
sns.boxplot(coffee, x='Coffee_2019', whis=(0,100))
sns.stripplot(coffee, x='Coffee_2019')
plt.savefig('i/ex_02.png')
```

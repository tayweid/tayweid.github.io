```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-2-4/data/'
```

```python
# Load Data
data = pd.read_csv(file_path + 'Starbucks_Promos.csv', index_col=0)
data.head()
```

## Revenue by Offer ID

```python
data.groupby('Offer ID')['Revenue'].sum()
```

```python
offer_rev = data.groupby('Offer ID')['Revenue'].sum()
offer_rev.plot(kind='barh')
```

## Redemption Rate

```python
# transactions / offers

transactions = data[data['Event'] == 'transaction']
tcount = transactions.groupby('Offer ID')['Event'].count()

offers = data[data['Event'] == 'offer']
ocount = offers.groupby('Offer ID')['Event'].count()

redemption_rate = tcount / ocount
redemption_rate
```

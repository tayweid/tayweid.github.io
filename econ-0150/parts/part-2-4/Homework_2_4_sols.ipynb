```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-2-4/data/'
```

## Q1. Amazon Book Sales

We'll explore the list of 50 bestselling Amazon books each year between 2009 and 2021. Prices are expressed in US dollars and rounded to the nearest dollar.

```python
books = pd.read_csv(file_path + 'amazon_book_sales.csv', index_col=0)
books.head()
```

#### a) Create a figure that will display the count of fiction books for each year.

```python
fiction = books[books['Genre'] == 'Fiction']
fic_year = fiction.groupby('Year')['Genre'].count()
fic_year
```

```python
fic_year.plot(kind='barh')
plt.title('Count of Fiction Books by Year')
plt.xlabel('')
plt.ylabel('')
sns.despine()
```

## Q2. Marriage Rates by Continent

The following questions are based on **crude marriage rates** — numbers of marriages per one thousand inhabitants — in 1990 and 2019. Each row represents a different country.


```python
marriage = pd.read_csv(file_path + 'marriage_rates_by_continent.csv')
marriage.head()
```

a) Create a figure showing the mean marriage rate by continent.

```python
mmbc = marriage.groupby('Continent')['Marriage_Rate_per_1000'].mean()
mmbc.plot(kind='barh')
plt.title('Mean Marriage Rate by Continent')
plt.xlabel('')
plt.ylabel('')
sns.despine()
```

b) Which continent has the lowest mean marraige rate?

South America

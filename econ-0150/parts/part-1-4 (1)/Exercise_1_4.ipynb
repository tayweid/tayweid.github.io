# Exercise 1.4 | Panel Data (Long Format)

```
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-1-4/data/'
```

## Coffee Shop Transactions

Use `Coffee_Sales_Reciepts.csv` to help inform where to hire a barista.

This is **long format panel data**: each transaction is a row, and a column identifies the shop.

```
# Load Dataset
sales = pd.read_csv(file_path + 'Coffee_Sales_Reciepts.csv', index_col=0)
sales.head()
```

### Bar Chart: Which shop is busiest?

```
# Bar graph
sns.countplot(sales, x='Shop', hue='Shop')
plt.title('Transactions by Shop')
```

### Histogram: What time of day is busiest?

```
# Histogram
sns.histplot(sales, x='Hours', bins=range(0, 24, 1))
plt.xlabel('Hour of the Day')
plt.title('Transactions by Hour')
```

### Multi-Histogram: Breaking out by shop

```
# Multi-histogram
sns.histplot(sales, x='Hours', hue='Shop', bins=range(0, 24, 1))
plt.xlabel('Hour of the Day')
plt.title('Transactions by Shop')
```

### Faceted Histogram: Each shop gets its own panel

```
# Faceted histogram
sns.displot(sales, x='Hours', col='Shop', bins=range(0, 24, 1))
```

### Count Hourly by Shop: Using groupby

```
# Create hourly counts by shop
hourly_counts = sales.groupby(['Shop', 'Hours']).size().reset_index(name='Count')
hourly_counts.head(10)
```

```
# Multi-Lineplot from grouped data
sns.lineplot(hourly_counts, x='Hours', y='Count', hue='Shop')
plt.xlabel('Hour of the Day')
plt.title('Transactions by Shop')
```

### Multiple Line Graph: Shortcut using histplot

```
# Multiple-Line Graph shortcut
sns.histplot(sales, x='Hours', hue='Shop', bins=range(0, 24, 1), element='poly', fill=False)
plt.xlabel('Hour of the Day')
plt.title('Transactions by Shop')
```

## Exercise 3.5 | Simplest GLM

Lets build the simplest statistical model.

```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
from scipy import stats

# Difference in wait times
waits = pd.DataFrame({
    'wait_difference': [1.3, 2.4, 2.2, 1.3, 3. , 2.3, 0.8, 2.7, 2. , 2.9, 1.7, 1.9, 2.2,
       2.1, 2.4, 1.7, 1.7, 1.9, 1.7, 2.5, 2.4, 2.2, 1.9, 1.9, 2.7, 1.8,
       1.8, 2. , 1.9, 2.1]
})
```

#### GLM Approach (default null: b=0)

```python
# Intercept-only model: y ~ 1
model = smf.ols('wait_difference ~ 1', data=waits).fit()
print(model.summary())
```

#### Traditional t-Test

```python
t_stat, p_value = stats.ttest_1samp(waits['y'], 0)
print(t_stat, p_value)
```

#### Setting The Null

```python
null = 1.8
waits['y_null'] = waits['y'] - null
```

## Exercise 3.5 | The Simplest GLM

Lets build the simplest statistical model.

```python
# Imports
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load and Show Data
wait_difference = [1.3, 2.4, 2.2, 1.3, 3. , 2.3, 0.8, 2.7, 2. , 2.9, 1.7, 1.9, 2.2,
       2.1, 2.4, 1.7, 1.7, 1.9, 1.7, 2.5, 2.4, 2.2, 1.9, 1.9, 2.7, 1.8,
       1.8, 2. , 1.9, 2.1]
data = pd.DataFrame({'wait_difference': wait_difference})
data.head()
```

## Exploratory Data Analysis

```python
# Visualize: Numerical
sns.histplot(x='wait_difference', data=data)
```

```python
# Visualize: Numerical
sns.stripplot(y='wait_difference', data=data)
```

## General Linear Model

```python
# Plot the model on a stripplot
sns.stripplot(y='wait_difference', data=data)
```

GLM Approach (default null: b=0)

```python
# Model: y ~ b
model = smf.ols('wait_difference ~ 1', data).fit().summary() # Default Null: 0
print(model.tables[1])
```

Traditional t-Test

```python
# t-Test
t_stat, p_value = stats.ttest_1samp(data['wait_difference'], 0) # Null: 0
print(t_stat, p_value)
```

## Exercise 3.5 | Simplest GLM

Lets build the simplest statistical model.

```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
from scipy import stats

# Difference in wait times
wait_difference = [1.3, 2.4, 2.2, 1.3, 3. , 2.3, 0.8, 2.7, 2. , 2.9, 1.7, 1.9, 2.2,
       2.1, 2.4, 1.7, 1.7, 1.9, 1.7, 2.5, 2.4, 2.2, 1.9, 1.9, 2.7, 1.8,
       1.8, 2. , 1.9, 2.1]
data = pd.DataFrame({'wait_difference': wait_difference})
```

```python
data.describe()
```

## GLM Approach (default null: b=0)

```python
# Model: y ~ b
model = smf.ols('wait_difference ~ 1', data).fit().summary() # Default Null: 0
print(model.tables[1])
```

## Traditional t-Test

```python
# t-Test
t_stat, p_value = stats.ttest_1samp(data['wait_difference'], 0) # Null: 0
print(t_stat, p_value)
```

## Q5. For each of the following variables, what is the variable type?

Choose from: Binary Categorical, Nominal Categorical, Ordinal Categorical, Discrete Numerical, Continuous Numerical

"What is your favorite color?"

**Nominal Categorical** - Colors are categories with no natural ordering.

"Approximately how many miles away from Pittsburgh is your hometown?"

**Continuous Numerical** - Distance can take any value (including decimals) within a range.

"What is your (primary) major?"

**Nominal Categorical** - Majors are categories with no natural ordering.

"How much did you like your statistics class?"

**Ordinal Categorical** - This is a rating scale 1-10 with a natural order but without meaningful intervals.

"How excited are you for this class?"

**Ordinal Categorical** - This is a rating scale 1-10 with a natural order but without meaningful intervals.

"Do you feel confident in Excel?"

**Binary Categorical** - Only two possible responses: Yes or No.

## Q6. Summarizing 'Do you feel confident in Excel?'

a) Appropriately visualize the variable 'Do you feel confident in Excel?'.

```python
# Visualize the binary categorical variable using a bar chart
sns.countplot(data, x='Do you feel confident in Excel?')
```

b) How many students in the class feel confident in Excel?

```python
# Create a table
data['Do you feel confident in Excel?'].value_counts()
```

51 students feel confident in Excel.

## Q7. Summarizing 'What is your major?'

a) Appropriately visualize the variable 'What is your major?'.

```python
# Visualize the nominal categorical variable with a horizontal bar chart
sns.countplot(data, y='What is your (primary) major?')
```

b) How many students in the dataset are economics majors?

```python
# Create a summary table
data['What is your (primary) major?'].value_counts()
```

37 students are economics majors.

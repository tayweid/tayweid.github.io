```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-0/data/'
```

```python
# Load Data
data = pd.read_csv(file_path + 'Fall_2025_Survey.csv')
data.head()
```

## Q2. Summarizing 'Do you feel confident in Excel?'

a) Appropriately visualize the variable 'Do you feel confident in Excel?'.

```python
# Visualize the binary categorical variable using a bar
sns.countplot(data, y='Do you feel confident in Excel?')

# Extra Styling (not required)
plt.grid(True)
sns.despine(bottom=False, left=True, right=True, top=True)
plt.xlabel('')
plt.ylabel('')
plt.title('Do you feel confident in Excel?')
```

b) How many students in the class feel confident in Excel?

From the figure it looks like the number is something like 70. But we can also just find the number.

```python
# Create a table
data['Do you feel confident in Excel?'].value_counts()
```

72 students feel confident in Excel.

## Q3. Summarizing 'What is your major?'

a) Appropriately visualize the variable 'What is your major?'.

```python
# Visualize the nominal categorical variable with a horizontal bar chart
sns.countplot(data, y='What is your (primary) major?')

# Extra Styling
plt.grid(True)
sns.despine(bottom=False, left=True, right=True, top=True)
plt.xlabel('')
plt.ylabel('')
plt.title('What is your (primary) major?')
```

b) How many students in the dataset are economics majors?

It looks like roughly 47 students are economics majors in the dataset. But here we can be more specific too.

```python
# Create a summary table
data['What is your (primary) major?'].value_counts()
```

There are 46 economics majors in the dataset.

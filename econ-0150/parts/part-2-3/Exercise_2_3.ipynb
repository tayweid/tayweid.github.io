# Exercise 2.3 | Filtering

This exercise demostrates how to:

1. Filter data by category
2. Filter numerical data

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-2-3/data/'
```

## Coffee Shop Hours

```python
# Load Dataset
hours = pd.read_csv(file_path + 'Starbucks_Location_Hours.csv', index_col=0)
hours.head()
```

```python
# Histogram
sns.histplot(hours, x='open', bins=range(0,25,1))
```

```python
us_hours = hours[(hours['COUNTRY_CODE'] == 'US')]
us_hours.head()
```

```python
# Histogram
sns.histplot(us_hours, x='open', bins=range(0,25,1))
```

```python
# OR |
# AND &
us_ca_hours = hours[(hours['COUNTRY_CODE'] == 'US') | (hours['COUNTRY_CODE'] == 'CA')]
```

```python
# Histogram
sns.histplot(us_ca_hours, x='open', bins=range(0,25,1))
```

```python
us_ca_hours[us_ca_hours['open'] < 7]
```

## Coffee Production and GDP

```python
gdp = pd.read_csv(file_path + 'Beans_GDP_2019.csv', index_col=0)
gdp.head()
```

```python
# Filter for countries in the Americas
americas = gdp[gdp['continent'] == 'Americas']

# Scatterplot of Americas only
sns.scatterplot(americas, x='GDP', y='coffee_prod')
plt.title('Coffee Production vs GDP (Americas)')
plt.grid()
```

## Encoding Categories

We can use **color** and **shape** to distinguish categorical groups in a scatterplot.

```python
# Color Encoding by continent
sns.scatterplot(gdp, x='GDP', y='coffee_prod', hue='continent', s=60)
plt.title('Coffee Production vs GDP (Color = Continent)')
plt.grid()
```

```python
# Shape Encoding by continent
sns.scatterplot(gdp, x='GDP', y='coffee_prod', style='continent', s=60)
plt.title('Coffee Production vs GDP (Shape = Continent)')
plt.grid()
```

```python
# Color AND Shape by continent
sns.scatterplot(gdp, x='GDP', y='coffee_prod', hue='continent', style='continent', s=60)
plt.title('Coffee Production vs GDP (Color & Shape = Continent)')
plt.grid()
```

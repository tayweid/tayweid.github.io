# Exercise 1.3 | Time Series

This exercise demonstrates how to:
1) Create line plots to visualize trends over time
2) Transform nominal prices to real prices (inflation adjustment)
3) Create multi-boxplots to examine seasonality

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-1-3/data/'
```

## Line Plots

Use line plots to visualize how coffee prices change over time.

```python
# Load coffee prices data
prices = pd.read_csv(file_path + 'Monthly_Coffee_Prices.csv', index_col=0)
prices.head()
```

```python
# Convert date column to datetime
prices['date'] = pd.to_datetime(prices['date'], format='%Y-%m')
prices.head()
```

```python
# Create a line plot of coffee prices over time
sns.lineplot(prices, x='date', y='price')
plt.xlabel('')
plt.ylabel('Coffee Price ($)')
plt.title('Coffee Prices Over Time')
```

## Real Price Conversion

```python
cpi = pd.read_csv(file_path + 'Coffee_Prices_CPI.csv', index_col=0)
cpi.head()
```

```python
cpi['real_price'] = cpi['price'] / cpi['today']
```

## Multi-Boxplots for Seasonality

Use multi-boxplots to compare the distribution of coffee prices across months.

```python
# Load coffee prices and add month names
prices = pd.read_csv(file_path + 'Monthly_Coffee_Prices.csv', index_col=0)
prices['date'] = pd.to_datetime(prices['date'], format='%Y-%m')

# Create month name column
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_names = dict(zip(range(1, 13), months))
prices['month_name'] = prices['date'].dt.month.map(month_names)
prices.head()
```

```python
# Create a multi-boxplot of coffee prices by month
sns.boxplot(prices, x='price', y='month_name', 
            order=months, color='lightgrey', whis=(0, 100))
plt.xlabel('Coffee Price ($)')
plt.ylabel('')
plt.title('Coffee Price Distribution by Month')
```

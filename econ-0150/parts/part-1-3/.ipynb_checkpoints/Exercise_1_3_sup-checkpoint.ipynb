# Exercise 1.4 | Data Transformations

This exercise demonstrates how to:
1) Transform nominal prices to real prices
2) Transform totals to per capita values
3) Convert from linear to log scales

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-1-4/data/'
```

## Real Price Conversion

```python
cpi = pd.read_csv(file_path + 'Coffee_Prices_CPI.csv', index_col=0)
cpi.head()
```

```python
cpi['real_price'] = cpi['price'] / cpi['today']
```

## Per Capita Transformation

```python
# Load GDP and Population data
gdp_pop = pd.read_csv(file_path + 'GDP_Population_2019.csv')
gdp_pop.head()
```

```python
# Calculate GDP per capita
# GDP is in trillions, Population is in millions
# To get dollars per person: (GDP * 1e12) / (Population * 1e6) = GDP * 1e6 / Population
gdp_pop['GDP_per_capita'] = gdp_pop['GDP'] * 1e6 / gdp_pop['Population']
```

```python
# Compare the top countries by total GDP vs GDP per capita
gdp_pop.sort_values('GDP_per_capita', ascending=False).head(10)
```

## Convert Scale from Linear to Log

```python
gdp = pd.read_csv(file_path + 'Beans_GDP_2019.csv', index_col=0)
gdp.head()
```

```python
gdp['log_coffee_prod'] = np.log(gdp['coffee_prod'])
```

```python
gdp['log_GDP'] = np.log(gdp['GDP'])
```

```python
sns.scatterplot(gdp, x='log_coffee_prod', y='log_GDP')
```

```python
# Use a log scale without transforming the variable
sns.scatterplot(gdp, y='coffee_prod', x='GDP')
plt.xscale('log')
plt.yscale('log')
```

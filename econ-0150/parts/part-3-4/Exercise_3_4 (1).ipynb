# Exercise 3.4 | Hypothesis Tests

```python
import numpy as np
from scipy import stats
```

```python
s = 2.5
n = 30
se = s / np.sqrt(30)
xbar = 10.85
null = 10
```

## Confidence Intervals

```python
lower, upper = stats.t.interval(0.95, df=30, loc=null, scale=se)
```

```python
[lower.item(), upper.item()]
```

## p-values

```python
stats.t.cdf(xbar, df=n-1, loc=null, scale=se).item()
```

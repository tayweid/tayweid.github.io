## Part 2: Choosing the Best Prediction

The simplest model predicts every observation with a single number $b$:

$$\hat{y}_i = b$$

The error for each observation is $\epsilon_i = y_i - b$. We measure the model's total "wrongness" with **Mean Squared Error**:

$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - b)^2$$

**Question 3.** Compute the MSE when $b = 2.5$.

*Hint: `np.mean((data['wait_difference'] - b)**2)`*

**Question 4.** Compute the MSE when $b = 1.5$.

**Question 5.** Compute the MSE when $b = \bar{y}$ (the sample mean). Which of the three values of $b$ gives the smallest MSE?

**Question 6.** The code below computes MSE for many values of $b$ and plots the result. Run it. Where is the minimum? What shape is the curve?

```python
b_values = np.linspace(0, 4, 200)
mse_values = [np.mean((data['wait_difference'] - b)**2) for b in b_values]

plt.figure(figsize=(8, 4))
plt.plot(b_values, mse_values, 'b-', linewidth=2)
plt.axvline(data['wait_difference'].mean(), color='r', linestyle='--',
            label=f'Sample Mean = {data["wait_difference"].mean():.2f}')
plt.xlabel('b')
plt.ylabel('MSE')
plt.title('Mean Squared Error vs. b')
plt.legend()
sns.despine()
plt.show()
```

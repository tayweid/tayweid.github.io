```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# File Path
file_path = 'https://tayweid.github.io/econ-0150/parts/part-2-2/data/'
```

## Q1. Amazon Book Sales by Genre

We'll explore the list of 50 bestselling Amazon books each year between 2009 and 2021. The column `Genre` describes whether the bestseller is Fiction or Non-Fiction.

```python
books = pd.read_csv(file_path + 'amazon_book_sales.csv', index_col=0)
books.head()
```

#### a) Create a boxplot showing the distribution of `User Rating` by `Genre`.

```python
sns.boxplot(data=books, x='Genre', y='User Rating', whis=(0,100))
plt.title('User Rating by Genre')
```

#### b) Add a stripplot on top of your boxplot to see individual books.

```python
sns.boxplot(data=books, x='Genre', y='User Rating', whis=(0,100))
sns.stripplot(data=books, x='Genre', y='User Rating', alpha=0.3, color='black')
plt.title('User Rating by Genre')
```

#### c) Calculate the mean, standard deviation, and count of `User Rating` by `Genre`.

```python
books.groupby('Genre')['User Rating'].agg(['mean', 'std', 'count'])
```

#### d) Which genre has higher average ratings? Is there substantial overlap between the two distributions?

Fiction has a slightly higher average rating. Yes, there is substantial overlap between the two distributions  the boxplots show that most of the interquartile ranges overlap, and the stripplot reveals many individual books from both genres at similar rating levels.

## Q2. Marriage Rates by Continent

The following questions are based on **crude marriage rates**  numbers of marriages per one thousand inhabitants. Each row represents a different country.

```python
marriage = pd.read_csv(file_path + 'marriage_rates_by_continent.csv')
marriage.head()
```

#### a) Create a boxplot showing the distribution of `Marriage_Rate_per_1000` by `Continent`.

```python
sns.boxplot(data=marriage, y='Continent', x='Marriage_Rate_per_1000', whis=(0,100))
plt.title('Marriage Rate by Continent')
```

#### b) Add a stripplot to see individual countries within each continent.

```python
sns.boxplot(data=marriage, y='Continent', x='Marriage_Rate_per_1000', whis=(0,100))
sns.stripplot(data=marriage, y='Continent', x='Marriage_Rate_per_1000', alpha=0.3, color='black')
plt.title('Marriage Rate by Continent')
```

#### c) Calculate the mean, standard deviation, and count of marriage rates by continent.

```python
marriage.groupby('Continent')['Marriage_Rate_per_1000'].agg(['mean', 'std', 'count'])
```

#### d) Which continent has the lowest mean marriage rate? Which has the most variation (highest standard deviation)?

South America has the lowest mean marriage rate. Asia has the most variation (highest standard deviation).

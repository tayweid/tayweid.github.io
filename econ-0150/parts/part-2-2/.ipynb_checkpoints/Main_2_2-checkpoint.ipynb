## NEW: Distribution Visualizations by Category (Log2 Revenue)

These figures support concept_2_2_new.qmd which uses Starbucks Promos data with log base 2 transformed revenue to show spending distributions by offer type, setting up for t-test preparation in Part 3. Each unit increase represents a doubling of spending.

```python
# Load Data and apply log base 2 transform
data = pd.read_csv('data/Starbucks_Promos.csv')
transactions = data[data['Event'] == 'transaction'].copy()
transactions['log2_Revenue'] = np.log2(1 + transactions['Revenue'])

# Side-by-side boxplots by offer type
plt.figure(figsize=(10, 5))
sns.boxplot(data=transactions, x='Offer ID', y='log2_Revenue', 
            hue='Offer ID', palette=category_color_pal, legend=False, whis=(0,100))

# Style
plt.title('Log2 Spending Distribution by Offer Type')
plt.xlabel('')
plt.ylabel('Log2(1+Revenue)')
sns.despine(top=True, right=True)
plt.tight_layout()

# Save Figure
plt.savefig('i/box_01.png')
```

```python
# Violin plot by offer type
plt.figure(figsize=(10, 5))
sns.violinplot(data=transactions, x='Offer ID', y='log2_Revenue', 
               hue='Offer ID', palette=category_color_pal, legend=False)

# Style
plt.title('Log2 Spending Distribution by Offer Type')
plt.xlabel('')
plt.ylabel('Log2(1+Revenue)')
sns.despine(top=True, right=True)
plt.tight_layout()

# Save Figure
plt.savefig('i/violin_01.png')
```

```python
# Stripplot by offer type (sample for readability)
plt.figure(figsize=(10, 5))
sample = transactions.sample(n=min(3000, len(transactions)), random_state=42)
sns.stripplot(data=sample, x='Offer ID', y='log2_Revenue', 
              hue='Offer ID', palette=category_color_pal, alpha=0.4, jitter=True, legend=False)

# Style
plt.title('Individual Transactions by Offer Type')
plt.xlabel('')
plt.ylabel('Log2(1+Revenue)')
sns.despine(top=True, right=True)
plt.tight_layout()

# Save Figure
plt.savefig('i/strip_01.png')
```

```python
# Combined boxplot + stripplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=transactions, x='Offer ID', y='log2_Revenue', 
            hue='Offer ID', palette=category_color_pal, legend=False, whis=(0,100))
sample = transactions.sample(n=min(2000, len(transactions)), random_state=42)
sns.stripplot(data=sample, x='Offer ID', y='log2_Revenue', 
              alpha=0.2, color='black', jitter=True)

# Style
plt.title('Log2 Spending Distribution by Offer Type')
plt.xlabel('')
plt.ylabel('Log2(1+Revenue)')
sns.despine(top=True, right=True)
plt.tight_layout()

# Save Figure
plt.savefig('i/box_strip_01.png')
```

```python
# Compare two offer types (Bogo 5 vs 5off20)
two_offers = transactions[transactions['Offer ID'].isin(['Bogo 5', '5off20'])]

plt.figure(figsize=(8, 5))
sns.boxplot(data=two_offers, x='Offer ID', y='log2_Revenue', 
            hue='Offer ID', palette=category_color_pal, legend=False, whis=(0,100))
sample = two_offers.sample(n=min(2000, len(two_offers)), random_state=42)
sns.stripplot(data=sample, x='Offer ID', y='log2_Revenue', 
              alpha=0.3, color='black', jitter=True)

# Style
plt.title('Comparing Two Offer Types')
plt.xlabel('')
plt.ylabel('Log2(1+Revenue)')
sns.despine(top=True, right=True)
plt.tight_layout()

# Save Figure
plt.savefig('i/compare_01.png')
```

```python
# Compare two offer types with overlap highlighted
two_offers = transactions[transactions['Offer ID'].isin(['Bogo 5', '5off20'])]

plt.figure(figsize=(8, 5))
sns.boxplot(data=two_offers, x='Offer ID', y='log2_Revenue', 
            hue='Offer ID', palette=category_color_pal, legend=False, whis=(0,100))
sample = two_offers.sample(n=min(2000, len(two_offers)), random_state=42)
sns.stripplot(data=sample, x='Offer ID', y='log2_Revenue', 
              alpha=0.3, color='black', jitter=True)

# Highlight overlap region
overlap_min = two_offers.groupby('Offer ID')['log2_Revenue'].quantile(0.25).max()
overlap_max = two_offers.groupby('Offer ID')['log2_Revenue'].quantile(0.75).min()
plt.axhspan(overlap_min, overlap_max, alpha=0.2, color='gray', label='Overlap region')

# Style
plt.title('Overlap Between Distributions')
plt.xlabel('')
plt.ylabel('Log2(1+Revenue)')
sns.despine(top=True, right=True)
plt.tight_layout()

# Save Figure
plt.savefig('i/compare_02.png')
```

```python
# Average log2 spending by offer type (bar chart)
plt.figure(figsize=(10, 4))
group_means = transactions.groupby('Offer ID')['log2_Revenue'].mean().sort_values()
group_means.plot(kind='barh', color=category_color_pal[:len(group_means)])

# Style
plt.title('Average Log2 Spending by Offer Type')
plt.xlabel('Mean Log2(1+Revenue)')
plt.ylabel('')
sns.despine(top=True, right=True)
plt.tight_layout()

# Save Figure
plt.savefig('i/i_08.png')
```

```python
# Log transformation visualization: jointplot with marginal histograms
# Shows how the transformation spreads out the skewed distribution

data = pd.read_csv('data/Starbucks_Promos.csv')
transactions = data[data['Event'] == 'transaction'].copy()
transactions['log2_Revenue'] = np.log2(1 + transactions['Revenue'])

# Sample for readability
sample = transactions.sample(n=min(3000, len(transactions)), random_state=42)

# Create jointplot
g = sns.jointplot(data=sample, x='Revenue', y='log2_Revenue', 
                  kind='scatter', alpha=0.3, height=6)

# Add reference points
ref_points = [7, 15, 31, 63]
for pt in ref_points:
    g.ax_joint.axvline(x=pt, color='red', linestyle='--', alpha=0.5, linewidth=1)
    g.ax_joint.axhline(y=np.log2(1+pt), color='red', linestyle='--', alpha=0.5, linewidth=1)

# Labels
g.ax_joint.set_xlabel('Revenue ($)')
g.ax_joint.set_ylabel('Log2(1 + Revenue)')

plt.tight_layout()
plt.savefig('i/log_transform_joint.png', dpi=150, bbox_inches='tight')
```

# Exercise 3.1 | Random Variables and Probability Functions

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# County A Sleep Times: Normal Distribution
mu = 7.2
sigma = 1.5
```

## Q1. What proportion of the population sleeps less than 5 hours?

```python
stats.norm.cdf(5, loc=mu, scale=sigma).item()
```

## Q2. What proportion of the population sleeps more than 9 hours?

```python
1 - stats.norm.cdf(9, loc=mu, scale=sigma).item()
```

## Q2. How much sleep does the middle 92% of the population get?

```python
lower_bound = stats.norm.ppf(0.04, loc=mu, scale=sigma).item()
upper_bound = stats.norm.ppf(0.96, loc=mu, scale=sigma).item()
```

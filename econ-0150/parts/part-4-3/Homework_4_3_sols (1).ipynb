## Q1. Mental Health and Income Group

```python
data = pd.read_csv(file_path + 'BRFSS_cleaned.csv')
data.head()
```

```python
data['INCOME2'].unique()
```

```python
data = data.dropna()
```

#### a) Run a linear regression such that the mental health variable is the outcome variable and the indicator for income less than \$10000 is the independent variable. Interpret your results.

```python
data['under_10k'] = (data['INCOME2'] == 1).astype(int)
data['under_10k']
```

```python
# Model: MENTHLTH = b + m * under_10k
model = smf.ols('MENTHLTH ~ under_10k', data=data).fit()
print(model.summary().tables[1])
```

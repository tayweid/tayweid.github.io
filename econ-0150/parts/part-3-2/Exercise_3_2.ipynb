# Exercise 3.2 | Sampling and the Central Limit Theorem

In this exercise we generate samples and simulate the central limit theorem.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def sample(n_rolls, n_samples=1):
    # Dice rolling (uniform distribution from 1-6)
    return np.random.randint(1, 7, size=n_rolls).tolist()

def samples(n_rolls, n_samples):
    # Dice rolling (uniform distribution from 1-6)
    return [np.mean(sample(n_rolls)).mean() for _ in range(n_samples)]
```

```python
import matplotlib.pyplot as plt

# Create the figure and axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 4))

# First bar chart (P(Home scores 0) and P(Home scores 1))
p_home_0 = 0.24
p_home_1 = 0.32
p_home_empty = 1 - (p_home_0 + p_home_1)

# Bar for home scores
ax1.barh(0, p_home_0 * 100, color='#4169E1FF', edgecolor='black', height=0.6)
ax1.barh(0, p_home_1 * 100, left=p_home_0 * 100, color='#4169E199', edgecolor='black', height=0.6)
ax1.barh(0, p_home_empty * 100, left=(p_home_0 + p_home_1) * 100, color='white', edgecolor='black', height=0.6)

# Remove axes for the first chart
ax1.set_xlim(-1, 101)
ax1.set_ylim(-0.5, 0.5)
ax1.axis('off')

# Add labels for the first chart
ax1.text(p_home_0 * 50, 0.45, 'P(Home scores 0)', ha='center', va='center', fontsize=14, color='#4169E1FF')
ax1.text(p_home_0 * 100 + p_home_1 * 50, 0.45, 'P(Home scores 1)', ha='center', va='center', fontsize=14, color='#4169E1FF')

# Add lines below the text for the first chart
ax1.plot([0, p_home_0*0.98 * 100], [0.4, 0.4], color='#4169E1FF', linewidth=2)
ax1.plot([p_home_0*1.02 * 100, (p_home_0 + p_home_1) * 100], [0.4, 0.4], color='#4169E199', linewidth=2)

# Second bar chart (P(Home scores 0) and P(Away scores 0))
p_home_0_2 = 0.24
p_away_0 = 0.33
p_away_overlap = 0.07  # Overlapping region
p_away_empty = 1 - (p_home_0_2 + p_away_0)

# Bar for home scores in the second chart
ax2.barh(0, (p_home_0_2 - p_away_overlap) * 100, color='#4169E1FF', edgecolor='black', height=0.6)

# Bar for away scores in the second chart (without hatching)
ax2.barh(0, (p_away_0 - p_away_overlap) * 100, left=(p_home_0_2) * 100, color='tomato', edgecolor='black', height=0.6, zorder=2)

# Overlapping region (drawn on top with transparency)
ax2.barh(0, p_away_overlap * 100, left=(p_home_0_2 - p_away_overlap) * 100, color='tomato', edgecolor='#4169E1FF', height=0.6, hatch='///////', zorder=-3)

# Bar for the remaining empty space in the second chart
ax2.barh(0, (p_away_empty+ p_away_overlap) * 100, left=(p_home_0_2 + p_away_0 - p_away_overlap) * 100, color='white', edgecolor='black', height=0.6)

# Remove axes for the second chart
ax2.set_xlim(-1, 101)
ax2.set_ylim(-0.5, 0.5)
ax2.axis('off')

# Add labels for the second chart
ax2.text(p_home_0_2 * 50, 0.4, 'P(Home scores 0)', ha='center', va='center', fontsize=14, color='#4169E1FF')
ax2.text(p_home_0_2 * 100 + (p_away_0 - p_away_overlap) * 50, 0.55, 'P(Away scores 0)', ha='center', va='center', fontsize=14, color='tomato')

# Add lines below the text for the second chart
ax2.plot([0, (p_home_0_2) * 100], [0.35, 0.35], color='#4169E1FF', linewidth=2)
ax2.plot([(p_home_0_2 - p_away_overlap) * 100, (p_home_0_2 + p_away_0 - p_away_overlap) * 100], [0.5, 0.5], color='tomato', linewidth=2)

# Show the plot
plt.tight_layout()
plt.savefig('Figures/Part2_1_8.png')
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Define color parameters
win_color_1 = 'tomato'
draw_loss_color_1 = '#FFD6CC'
win_color_2 = 'royalblue'
draw_loss_color_2 = '#4169E133'

# Define height ratio for the bars
height_ratio = 0.05  # The first bar will take up 5% of the height, the second 95%

# Create the figure and axes
fig, ax = plt.subplots(1, 1, figsize=(6, 6))

# First bar chart (Red Card) - Win and Draw/Loss
p_win_1 = 0.24
p_loss_draw_1 = 1 - p_win_1

# Bar for Win
ax.barh(0, p_win_1 * 100, color=win_color_1, edgecolor='black', height=height_ratio)
# Bar for Draw/Loss
ax.barh(0, p_loss_draw_1 * 100, left=p_win_1 * 100, color=draw_loss_color_1, edgecolor='black', height=height_ratio)

# Second bar chart (No Red Card) - Win and Draw/Loss
p_win_2 = 0.46
p_loss_draw_2 = 1 - p_win_2

upp = 0.5
# Bar for Win
ax.barh(upp, p_win_2 * 100, color=win_color_2, edgecolor='black', height=1 - height_ratio)
# Bar for Draw/Loss
ax.barh(upp, p_loss_draw_2 * 100, left=p_win_2 * 100, color=draw_loss_color_2, edgecolor='black', height=1 - height_ratio)

# Add labels for the first bar (Red Card)
ax.text(p_win_1 * 50, - height_ratio/2 - 0.03, 'Win: 0.24', ha='center', va='center', fontsize=14, color=win_color_1)
ax.text(p_win_1 * 100 + p_loss_draw_1 * 50, - height_ratio/2 - 0.03, 'Draw/Loss: 0.76', ha='center', va='center', fontsize=14, color=win_color_1)

# Add labels for the second bar (No Red Card)
ax.text(p_win_2 * 50, 1/2 + (1-height_ratio)/2 + 0.03, 'Win: 0.46', ha='center', va='center', fontsize=14, color=win_color_2)
ax.text(p_win_2 * 100 + p_loss_draw_2 * 50, 1/2 + (1-height_ratio)/2 + 0.03, 'Draw/Loss: 0.54', ha='center', va='center', fontsize=14, color=win_color_2)

# Add side labels
ax.text(-6, 0, 'Red Card', fontsize=14, color=win_color_1, va='center', ha='left', rotation=90)
ax.text(-6, upp, 'No Red Card', fontsize=14, color=win_color_2, va='center', ha='left', rotation=90)

# Add side labels
ax.text(101, 0, height_ratio, fontsize=14, color=win_color_1, va='center', ha='left', rotation=90)
ax.text(101, 0+upp, 1-height_ratio, fontsize=14, color=win_color_2, va='center', ha='left', rotation=90)

# Adjust limits and remove axes for the chart
ax.set_xlim(-1, 101)
#ax.set_ylim(-0.1, 1)
ax.axis('off')

# Display the plot
plt.tight_layout()
plt.savefig('Figures/Part2_1_20.png')
```

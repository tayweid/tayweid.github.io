## Question 1
---
*Consider the rating data set, which contains information on surveyed customer satisfaction from 12 different call center agents. 

From the raw data agent id <u>( ? )</u>  has the worst average ratings, while agent id <u>( ? )</u> has the best ratings.*

**Answer:**
Here I first generate some summary averages by agent id:

```r
agent.outcomes <- matrix(0, nrow=12, ncol=10)
for (i in 1:12){
    df <- subset(rating.data,agent.id==i )
    agent.outcomes[i , ] <- c(i, mean(df$rating.val),
                          median(df$rating.val),
                         mean(df$rating.val==1),
                         mean(df$rating.val==2),
                         mean(df$rating.val==3),
                         mean(df$rating.val==4),
                         mean(df$rating.val==5),
                         mean(df$type=="corporate"),
                         mean(df$call.duration)
                        )
}
colnames(agent.outcomes) <- c("Agent","Mean.rating","Median.rating","Frac.1","Frac.2","Frac.3","Frac.4","Frac.5","Frac.Corp","Mean.duration")
agent.stats <- data.frame(agent.outcomes)
round(agent.stats[order(agent.stats$Mean.rating),],2)
```

So, looking at the data sorted by the mean rating, the best agent is **7** and the worst agent is **10**. However, you can see here that agents 4, 9 and 10 all face much longer calls, and a greater share of corporate callers.

## Question 2
---
*Run both an ordered logit regression and an ordered probit for the satisfaction ratings on:*
* a dummy variable indicating the type of caller (i.e whether they are a corporate customer or not)
* the log of the call duration
* A factor dummy for each agent id.

*Which model out of the Probit of Logit better fits the data?*

**Answer:**

```r
library(MASS)
ologit.est <- polr(as.factor(rating.val)~log(call.duration)+as.factor(agent.id)+as.factor(type),data=rating.data)
oprobit.est <- polr(as.factor(rating.val)~log(call.duration)+as.factor(agent.id)+as.factor(type),data=rating.data,method="probit")
round(c("Logit.AIC"=AIC(ologit.est),"Probit.AIC"=AIC(oprobit.est)),2)
```

So, based off of the Akaike information criterion the **Probit** is the better fitting model here. (When I generated this data I did use a normal, so this is good!)

## Question 5
---
*Again using the better-fitting model, calculate the probability Agent-7 has a "Very Bad" or a "Bad" rating for a 5 minute non-corporate caller?*

**Answer:**

The better fitting model is probit model, so we can either calculate this directly to get the linear predictor:

```r
beta.probit <- coef(oprobit.est)
predVal <- unname(beta.probit['log(call.duration)']*log(5)+beta.probit['as.factor(type)non.corporate']+beta.probit['as.factor(agent.id)7'])
round(predVal,3)
```

And we can then use the normal distribution to figure out the probability of being below the "Bad"/"Neutral" threshold:

```r
c("Prob.VBad.Bad"=round(unname(pnorm(oprobit.est$zeta["2|3"]-predVal)),3)) 
```

Alternatively we can just make a new data frame to predict the probabilities:

```r
pred.df <- data.frame(type="non.corporate",call.duration=5,agent.id=7)
predicted.probs <- predict(oprobit.est, newdata=pred.df,type="probs")
round(predicted.probs,3)
```

And we get the same answer by summing the first two ratings!

```r
predicted.probs[1]+predicted.probs[2]
```

## Question 6
---
*Three of the agents (4, 9 and 10) work on the second-tier of support, dealing with more-complicated problems. Using their call lengths, and caller types, predict the average rating Agent-7 would have if she faced the second-tier caller types? *

**Answer:**

Here we will again use the predict function to get the outcomes for agent 7 if facing the call types from agents 4,9 and 10. First let's get these agents calls, and replace their ID with 7:

```r
pred.data.7 <- subset(rating.data,agent.id==4 | agent.id==9 | agent.id==10)
pred.data.7['agent.id'] <- 7
```

Then we can use this data to generate the predicted probabilities:

```r
predicted.probs.7 <- predict(oprobit.est, newdata=pred.data.7,type="probs")
head(predicted.probs.7)
```

So we want to use these probabilities to capture the *expected rating*, so we'll just multiply each row by (1,2,3,4,5) to get the expected rating for each call: (here I use a matrix multiplication to do this) 

```r
expected.rating <- predicted.probs.7 %*% c(1,2,3,4,5)
head(expected.rating)
```

And I can then just take averages across all of the calls to get the expected rating if agent 7 faced the second-tier's calls:

```r
round( mean(expected.rating)
      ,2)
```

So a 2.73, which is far below the average 3.9 rating agent 7 gets on their other calls, but which is still good compared to the ratings that 4,9 and 10 get!

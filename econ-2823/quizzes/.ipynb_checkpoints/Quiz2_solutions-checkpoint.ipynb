## Question 1:
---
Suppose that you have 5 data points (1, 2 ,5, 3, 2 ,1) where the density is linearly decreasing on the interval , given by:
$$f(x)=\begin{cases} \frac{2(\theta-x)}{\theta^2} & \text{if } u\in [0,\theta]
,\\ 0 & \text{otherwise.} \end{cases} $$

 What is the maximum-likelihood estimator for $\theta$ (to 3 decimal places)?

### Answer:
Analytical part:
 We can write out the log-likelihood of the data as:
 $$l(\theta)=\sum_i \log \left(\frac{2(\theta-x_i)}{\theta^2} \right) = \sum_i \left[ \log(2)+\log(\theta-x_i) -2\log(\theta) \right]$$
 so long as $\theta$ is larger than the maximum in out data, otherwise the log-likelihood is $-\infty$!

The log-likelihood for $\theta>5$ is therefore:
$$ N \log(2)+\sum^N_{i=1} \log(\theta-x_i)  -2N\log(\theta)$$

Let's quickly graph this to make sure we know what it looks like:

```r
log.like.q1 <- function(theta) 6*log(2)+2*log(theta-1)+2*log(theta-2)+log(theta-3)+log(theta-5)-2*6*log(theta)
base + geom_function( fun =log.like.q1, colour=Pitt.Gold, size=3)+
xlim(5.2, 10)+xlab("theta")+ylab("log.likelihood")
```

One way of getting the answer here quite esily is just maximizing the log-likelihood across the range $[5.2,10]$ in the above plot, which is easy enough to do with:

```r
optim.out <- optimize(log.like.q1, interval=c(5.2,10),maximum=TRUE)
optim.out
round(optim.out$maximum,3)
```

Which gives us $\hat{\theta}=6.318$

Alternatively we can derive the first-order condition for an interior maximum with:
$$\frac{\partial l(\theta)}{\partial \theta}=\sum^N_{i=1} \frac{1}{\theta-x_i} - \frac{2N}{\theta} =0$$

Or, expanding the sum as we only have four values:
$$\frac{2}{\theta-1}+\frac{2}{\theta-2}+\frac{1}{\theta-3}+\frac{1}{\theta-5} -\frac{2N}{\theta}=0$$

But this is again a fairly non-linear equation where I'll struggle to find a root analytically. Graphing the function:

```r
d.log.like.q1 <- function(theta) 2/(theta-1)+2/(theta-2)+1/(theta-3)+1/(theta-5)-2*6/(theta)
base + geom_function( fun =d.log.like.q1, colour=Pitt.Gold, size=3)+
xlim(5.2, 10)+xlab("theta")+ylab("log.likelihood")
```

```r
library(rootSolve)
round(uniroot.all(d.log.like.q1,interval=c(5.2,10)),3)
```

So we get the same answer.

## Question 2
---
This one you have to do analytically, as the likelihood function ends up not having an interior max. Instead, you need to just think about it a little.

*Suppose that you have 5 data points (-1, 0, 1, 5 ,3)  where each is an iid draw from a uniform distribution over $[a,b]$ (so the distribution is given by):*
$$ f(u)=\begin{cases} \frac{1}{b-a} & \text{if } u\in [a,b],\\ 0 & \text{otherwise.} \end{cases}$$

### Answer:
Under the uniform the density for each potential value is identical so long as they are all in the interval $[a,b]$, otherwise one of the probabilities will be 0 and 0 times a positive probability is zero! The joint likelihood for the data for any $a$ and $b$ is therefore given by:
$$
L(a,b)=\begin{cases} \left(\frac{1}{b-a}\right)^N & \text{if all the data points are in } \in [a,b],\\ 0 & \text{otherwise.} \end{cases}
$$

This looks like this where the more gold, the greater the likelihood:

![Image](https://alistairjwilson.github.io/MQE_AW/i/Quiz2.png)

This is easier to see if we realize that the log-likelihood is effectively given by:
$$
l(a,b)=\begin{cases} -N\log(b-a) & \text{if all the data points are in } \in [a,b],\\ -\infty & \text{otherwise.} \end{cases}
$$

So we need to keep make $b-a$ as small as possible to make the non-infinite number as large as possible, but all the data points have to be inside of the interval.

This leads to the simple solution that:
* $\hat{a}=\min_i x_i$
* $\hat{b}=\max_i x_i$

The data is given by $(-1, 0, 1, 5 ,3)$, so that means the min and max are given by -1 and 5, respectively.

### Unbiased?
If we're just taking the min/max of the data, this does seem like it will be biased though, as the highest value from $N$ draws is almost certainly going to be less than the upper limit. Just to show you this is the case, consider  building up to ten-thousand draws from a uniform on $[0,1]$:

```r
unif.draws <- runif(10)
c(n=10,min=min(unif.draws),max=max(unif.draws))
unif.draws <- runif(100)
c(n=100,min=min(unif.draws),max=max(unif.draws))
unif.draws <- runif(1000)
c(n=1000,min=min(unif.draws),max=max(unif.draws))
unif.draws <- runif(10000)
c(n=10000,min=min(unif.draws),max=max(unif.draws))
```

While the numbers here will tend towards the true values of 0 and 1 in probability as $n\rightarrow\infty$ (that is, the estimators are consistent), we know that the min and the max have to be biased upwards and downwards! A more technical answer for this revolves around the expectation of the order statistic of the data being bounded away from the lower/upper interval.

## Question 3
---

*The geometric distribution is a discrete distribution that counts the number of independent draws until a specified event with probability  $p$ occurs (flipping a coin until a head occurs, rolling a die until you get a six, etc.). The probability mass function for a geometric random variable $X$ is given by:*
$$ \Pr\left\{X=x\right\}=(1-p)^{x-1}\cdot p\text{    for }x=1,2,\ldots $$

*Suppose you have 5 draws from a geometric distribution with parameter $p$  given by  (9, 16, 12, 3, 10). What is  the maximum-likelihood estimator for $p$?*

### Answer:
This one's a little more straightforward. The log-likelihood for any value for $p\in(0,1)$ is given by:
 $$l(p)=\log(1-p)(\sum_i x_i-N)+N\log(p)$$
 
 The first-order condition for the maxima is given by:
 $$\frac{\partial l(p)}{\partial p}=\frac{N}{p}-\frac{\sum_i x_i-N}{1-p}=0$$
 
 which simplifies down to:
 $$ (\sum_i x_i-N)\hat{p}=(1-\hat{p})N \Longrightarrow \hat{p}=\frac{N}{\sum_i x_i}$$

So our estimator is $\hat{p}=\frac{5}{9+16+12+3+10}=\frac{5}{50}=\tfrac{1}{10}$. Checking our arithmetic here:

```r
5/(9+16+12+3+10)
```

So the answer here is exactly 0.1!

We can also do this over the log-likelihood, where the one thing we have to be careful over is that the `dgeom()` probability given in R is over $0,1,2,,\ldots$ so we have to take away 1 from each of the values!

```r
log.like.q3 <- function(p)  dgeom(x=8,prob=p,log=TRUE)+dgeom(x=15,prob=p,log=TRUE) +dgeom(x=11,prob=p,log=TRUE) +dgeom(x=2,prob=p,log=TRUE) +dgeom(x=9,prob=p,log=TRUE) 
base + geom_function( fun =log.like.q3, colour=Pitt.Gold,size=3)+
xlim(0.01, 0.3)+xlab("p")+ylab("log.likelihood")
```

```r
opt.q3 <- optimize(log.like.q3 ,interval=c(0.01,0.3),maximum=TRUE)
round(opt.q3$maximum,3)
```

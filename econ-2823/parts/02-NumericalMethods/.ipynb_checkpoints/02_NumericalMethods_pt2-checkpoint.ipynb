# Numerical Methods (pt 2) | Optimization


As we're continuing from last time, let me just import and redefine some of the functions from before.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import optimize
from scipy.optimize import fsolve

PITT_BLUE = "#003594"
PITT_GOLD = "#FFB81C"
PITT_DGRAY = "#75787B"
PITT_GRAY = "#97999B"
PITT_LGRAY = "#C8C9C7"

# Numerical derivative functions from pt1
def derivative_1(f, x, eps=1e-6):
    """Centered difference first derivative."""
    return (f(x + eps) - f(x - eps)) / (2 * eps)

def derivative_2(f, x, eps=1e-3):
    """Centered difference second derivative."""
    return (f(x + eps) - 2*f(x) + f(x - eps)) / eps**2
```

## Optimization

The key insight is that at any optimum (maximum or minimum), the slope of the function is zero:

![The slope is zero at every optimum](https://alistairjwilson.github.io/MQE_AW/i/OptimalSlope.gif)

### Option 1: Use Analytical Derivatives

If you can write down the derivative analytically, just find its roots using `fsolve` (from pt1). Lets define two test functions:

$$ f(x)= 2 x^2 - 10 x + 3 $$
$$ g(x)=f(x)e^{-x/10} $$

We can (easily) find both derivatives analytically:

$$ f'(x) = 4 x - 10 $$
$$ g'(x) = - \frac{1}{10} f(x) e^{-x/10} + (4x - 10)e^{-x/10} $$

```python
# Define functions
def f(x):
    return 2*x**2 - 10*x + 3

def g(x):
    return f(x) * np.exp(-x/10)

# Analytical derivatives
def d_f(x):
    return 4*x - 10

def d_g(x):
    return -0.1 * np.exp(-0.1*x) * f(x) + np.exp(-0.1*x) * (4*x - 10)
```

```python
x = np.linspace(0, 5, 200)                                                    
                                                                            
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))                         
                                          
ax1.plot(x, f(x), color=PITT_BLUE, linewidth=2)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title("f(x) = 2x² - 10x + 3")

ax2.plot(x, d_f(x), color=PITT_BLUE, linewidth=2)
ax2.axhline(y=0, color=PITT_GOLD, linewidth=1)
ax2.set_xlabel('x')
ax2.set_ylabel("f'(x)")
ax2.set_title("f'(x) = 4x - 10")

plt.tight_layout()
plt.show()
```

```python
# Find optimal by solving f'(x) = 0
f_sol = fsolve(d_f, 1)[0]
print(f"x* = {f_sol}, f(x) = {f(f_sol)}")
```

```python
x = np.linspace(0, 40, 200)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))                         

ax1.plot(x, g(x), color=PITT_BLUE, linewidth=2)                               
ax1.set_xlabel('x')                         
ax1.set_ylabel('g(x)')
ax1.set_title("g(x) = (2x² - 10x + 3)exp(-x/10)")

ax2.plot(x, d_g(x), color=PITT_BLUE, linewidth=2)
ax2.axhline(y=0, color=PITT_GOLD, linewidth=1)
ax2.set_xlabel('x')
ax2.set_ylabel("g'(x)")
ax2.set_title("g'(x)")

plt.tight_layout()
plt.show()
```

```python
# Find both solutions using fsolve with different starting points
g_sol_1 = fsolve(d_g, 0)[0]
g_sol_2 = fsolve(d_g, 20)[0]

print(f"x* = {g_sol_1}, g(x) = {g(g_sol_1)}")
print(f"x* = {g_sol_2}, g(x) = {g(g_sol_2)}")
```

### Option 2: Fully Numeric Optimization

What if we don't have an analytical derivative at all? We can go fully numeric by using numerical approximations for both $f'(x)$ and $f''(x)$.

The core idea is to apply the Newton-Raphson to the first-order condition:

$$ x_{n+1} = x_n - \frac{f'(x_n)}{f''(x_n)} $$

Here we're essentially replacing $f$ with $f'$ in the Newton-Raphson gradient optimizer approach.

This is essentially what gradient-based optimizers like BFGS (Broyden–Fletcher–Goldfarb–Shanno, essentially a smarter version of Newton-Raphson) do under the hood. Let's see it once before handing off to `scipy.optimize.minimize`.

```python
def nr_optimize(f, x0, tolerance=1e-8, eps=1e-6, max_steps=50):
    """Newton-Raphson optimization using numerical derivatives only."""
    
    xx = x0
    for i in range(max_steps):
        f_1 = derivative_1(f, xx, eps)           # numerical f'(x)
        f_2 = derivative_2(f, xx, np.sqrt(eps))  # numerical f''(x)
        xx = xx - f_1 / f_2                      # NR step: x - f'(x)/f''(x)
        if abs(f_1) < tolerance:
            return xx
            
    print("Did not converge")
    return xx
```

```python
# Test on both functions — no analytical derivatives needed
print(f"f(x) minimum: x = {nr_optimize(f, 1):.5f}")
print(f"g(x) minimum: x = {nr_optimize(g, 0):.5f}")
print(f"g(x) maximum: x = {nr_optimize(g, 20):.5f}")
```

### Type of Optima: Maxima vs Minima

Both local minima and local maxima satisfy the $f^\prime(x)=0$ condition. We can figure out which type by checking the second derivative:

* For a **maximum**: $f^{\prime\prime}(x)<0$ (slope decreasing)
* For a **minimum**: $f^{\prime\prime}(x)>0$ (slope increasing)

```python
# Check second-order conditions at each solution
soc_1 = derivative_2(g, g_sol_1, 1e-5)
soc_2 = derivative_2(g, g_sol_2, 1e-5)

print(f"Solution 1 (x = {g_sol_1}): f''(x) = {soc_1}")
print(f"  {'LOCAL MINIMUM' if soc_1 > 0 else 'LOCAL MAXIMUM'}")

print(f"\nSolution 2 (x = {g_sol_2:.5f}): f''(x) = {soc_2:.5f}")
print(f"  {'LOCAL MINIMUM' if soc_2 > 0 else 'LOCAL MAXIMUM'}")
```

```python
# Visualize the solutions
fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(g_sol_1 - 5, g_sol_2 + 5, 200)

ax.axhline(y=0, color=PITT_GOLD, linewidth=1, linestyle='--', alpha=0.5)
ax.plot(x, g(x), color=PITT_BLUE, linewidth=2, label='g(x)')
ax.scatter([g_sol_1, g_sol_2], [g(g_sol_1), g(g_sol_2)], 
           color='black', s=100, zorder=5)

ax.set_xlabel('x')
ax.set_ylabel('g(x)')
plt.show()
```

## Global vs Local Optima

Only one of these is a **global** optimum. Let's zoom out to see the bigger picture.

```python
# Zoom out
fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(-5, 105, 500)

ax.plot(x, g(x), color=PITT_BLUE, linewidth=2)
ax.axhline(y=0, color=PITT_GOLD, linewidth=1, linestyle='--', alpha=0.5)
ax.scatter([g_sol_1, g_sol_2], [g(g_sol_1), g(g_sol_2)], color='black', s=100, zorder=5)

ax.set_xlabel('x')
ax.set_ylabel('g(x)')
plt.show()
```

## Local Solutions: The Challenge

When minimizing a function that is not single-peaked/troughed, gradient methods can get stuck at local solutions.

```python
# Define a wavy function with many local optima
def wavy_f(x):
    return (
        (-(x-2)**2) * np.cos(x-2) - 
        90 * np.cos(20*(x-2)) * np.exp(-30*(x-2)**4) + 
        np.exp(np.clip((x-2)**2, None, 500))
    )
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(0, 4, 500)
ax.plot(x, wavy_f(x), color=PITT_BLUE, linewidth=2)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(0, 4, 500)
ax.plot(x, wavy_f(x), color=PITT_BLUE, linewidth=2)

starts = [0, 1.6, 2.05, 2.1, 2.3]
for x0 in starts:
    result = optimize.minimize(wavy_f, x0, method='BFGS')
    ax.scatter([result.x[0]], [result.fun], color='black', s=100, zorder=5)

def anti_wavy(x):
    return - wavy_f(x)

for x0 in starts:
    result = optimize.minimize(anti_wavy, x0, method='BFGS')
    if result.x[0] > 0 and result.x[0] < 4:
        ax.scatter([result.x[0]], [wavy_f(result.x[0])], color=PITT_GOLD, s=100, zorder=5)

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
plt.show()
```

### Multi-dimensional Optimization

Consider the function:
$$ f(x_1,x_2)=2 x_1^2+2 x_2^2-2 x_1 x_2 - 5x_1+x_2$$

The first-order conditions give us:
$$\begin{array}{rccl} 
\frac{\partial f(x_1,x_2)}{\partial x_1} & = & 4 x_1- 2 x_2 -5&=0 \\
\frac{\partial f(x_1,x_2)}{\partial x_2} & = & 4 x_2- 2 x_1 +1 &=0 
\end{array}$$

The unique minimizing solution is: $x_1^\star=\tfrac{3}{2}$ and $x_2^\star=\tfrac{1}{2}$.

```python
# Define a function of 2 inputs
def f2(x):
    return 2*x[0]**2 + 2*x[1]**2 - 2*x[0]*x[1] - 5*x[0] + x[1]

print(f"f2([0, 0]) = {f2([0, 0])}")
print(f"f2([1.5, 0.5]) = {f2([1.5, 0.5])}")
print(f"f2([1.6, 0.4]) = {f2([1.6, 0.4])}")
```

### Multi-dimensional Second-Order Conditions (2 variables)

The multi-dimensional analog to the second-derivative test is the **Hessian matrix** of second-order partial derivatives. 

$$H = \begin{bmatrix} \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} \\ \frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} \end{bmatrix}$$

The diagonal entries tell you curvature along each axis individually. But the function could curve up along $x_1$ and up along $x_2$, but still curve *down* along some diagonal direction because of the cross-partial terms. 

For a $2 \times 2$ Hessian you can use the determinant test from calculus ($\det(H) > 0$ and $f_{x_1 x_1} > 0$ → minimum). 


### Multi-dimensional Second-Order Conditions (3+ variables$)

The multi-dimensional analog to the second-derivative test is the **Hessian matrix** of second-order partial derivatives. 

$$H = \begin{bmatrix} \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} \\ \frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} \end{bmatrix}$$

For 3+ variables that gets messy. Eigenvalues are the general version that scales to any dimension, and Python computes them for us with `np.linalg.eigvals(H)`:


Conveniently, BFGS returns an approximate inverse Hessian, so we can check this directly.

#### What are eigenvalues of the Hessian?


Eigenvalues decompose the Hessian into its **principal curvatures** — the curvature along the directions where things are simplest (no mixing between dimensions). If all principal curvatures are positive, the surface curves up in *every* direction.



* All eigenvalues **positive** → minimum
* All eigenvalues **negative** → maximum
* **Mixed** signs → saddle point
* An eigenvalue **near zero** → the surface is nearly flat in that direction (identification problem)

The second-order conditions become:

* **Positive definite** Hessian (all eigenvalues > 0) → local **minimum**
* **Negative definite** Hessian (all eigenvalues < 0) → local **maximum**

### Optimization in Python with SciPy

Python's `scipy.optimize` module provides powerful optimization routines.

**Key functions:**
| Function | Use Case |
|---|--------|
| `optimize.minimize(f, x0, method='BFGS')` | Multivariate minimization with gradient-based method |
| `optimize.minimize(f, x0, method='Nelder-Mead')` | Multivariate minimization without gradients |
| `optimize.minimize_scalar(f, bounds=(a,b), method='bounded')` | Univariate minimization on a bounded interval |

```python
# Optimize the 2D function using BFGS
result = optimize.minimize(f2, [0, 0], method='BFGS')

print("Optimization Result:")
print(f"  Optimal x: {result.x}")
print(f"  Optimal value: {result.fun}")
print(f"  Success: {result.success}")

# BFGS gives us the approximate inverse Hessian — invert to get the Hessian
hess = np.linalg.inv(result.hess_inv)
eigenvalues = np.linalg.eigvals(hess)

print(f"\nHessian:\n{hess}")
print(f"Eigenvalues: {eigenvalues}")
print(f"Positive definite (minimum)? {np.all(eigenvalues > 0)}")
```

### Local Solutions in Multiple Dimensions

The problem of getting stuck at local solutions doesn't go away with more dimensions!

```python
# 2D version of our wavy function
def wavy_f_2d(x):
    return (wavy_f(x[0]) + 90) * (1 + (x[1] - 3)**2) / 100

# BFGS may find a local minimum
result_bfgs = optimize.minimize(wavy_f_2d, [0, 0], method='BFGS')

print("BFGS result:")
print(f"  Found: {result_bfgs.x}")
print(f"  Value: {result_bfgs.fun}")

# Actual global minimum
print(f"\nGlobal minimum at [2, 3]: {wavy_f_2d([2, 3])}")
```

### Non-Gradient Methods: Nelder-Mead

The Nelder-Mead method does not use gradients at all. Instead, it creates a simplex (a shape with one more vertex than the number of dimensions) and iteratively adjusts it to find the minimum.

A nice visualization: https://www.benfrederickson.com/numerical-optimization/

Nelder-Mead can do poorly on high-dimensional problems (see the city-pair example at the visualization link above), or where there is flatness in the objective. It also performs poorly on one-dimensional problems. For those, use `minimize_scalar` with `method='bounded'` instead.

```python
# Nelder-Mead in Python
result_nm = optimize.minimize(wavy_f_2d, [0, 0], method='Nelder-Mead')

print("Nelder-Mead result:")
print(f"  Found: {result_nm.x}")
print(f"  Value: {result_nm.fun}")
```

```python
# For 1D optimization, use minimize_scalar
result_1d = optimize.minimize_scalar(g, bounds=(-10, 5), method='bounded')

print("1D optimization result:")
print(f"  Minimum at x = {result_1d.x:.6f}")
print(f"  Value: {result_1d.fun:.6f}")
```

```python
# Finding a local minimum in the wavy function
result_wavy = optimize.minimize_scalar(wavy_f, bounds=(1.5, 2.5), method='bounded')

print(f"Wavy function minimum: x = {result_wavy.x:.6f}")
print(f"Value: {result_wavy.fun:.6f}")
print(f"Compare to wavy_f(2) = {wavy_f(2):.6f}")
```

### Simulated Annealing and Global Optimization

When we're concerned about reaching global optima, **simulated annealing** is a stochastic method that can help escape local minima.

In scipy, we can use `optimize.dual_annealing()` or `optimize.basinhopping()`.

Here's an excellent visualization of how simulated annealing explores the solution space, accepting worse solutions early on to escape local optima, then gradually "cooling" to settle on the global optimum:

[Simulated Annealing Visualization (YouTube)](https://www.youtube.com/watch?v=iaq_Fpr4KZc)

```python
# Dual annealing for global optimization
result_da = optimize.dual_annealing(wavy_f_2d, bounds=[(-5, 10), (-5, 10)])

print("Dual Annealing result:")
print(f"  Found: {result_da.x}")
print(f"  Value: {result_da.fun}")
print(f"  Function evaluations: {result_da.nfev}")
```

```python
# Basin-hopping: combines local optimization with random restarts
result_bh = optimize.basinhopping(wavy_f_2d, [0, 0], niter=100)

print("Basin-Hopping result:")
print(f"  Found: {result_bh.x}")
print(f"  Value: {result_bh.fun}")
```

```python
# Combining approaches: global search followed by local refinement
result_global = optimize.dual_annealing(wavy_f_2d, bounds=[(-5, 10), (-5, 10)])
result_refined = optimize.minimize(wavy_f_2d, result_global.x, method='BFGS')

print("Combined approach:")
print(f"  After global search: {result_global.x}, value = {result_global.fun}")
print(f"  After refinement:    {result_refined.x}, value = {result_refined.fun}")
```

### Other Global Optimization Methods

Beyond simulated annealing, there are other stochastic methods for global optimization:

- **Genetic Algorithms**: Inspired by biological evolution, these maintain a "population" of candidate solutions that are combined ("crossbreed") and randomly modified ("mutated") over generations, with the fittest solutions surviving. SciPy provides `scipy.optimize.differential_evolution()` for this.

- **Particle Swarm Optimization**: Models a swarm of particles exploring the search space, sharing information about good regions.

These methods trade computational cost for robustness against local optima. In practice, many researchers use a two-stage approach: a global method to find the right "neighborhood," followed by a gradient-based method (like BFGS) to fine-tune the solution — exactly as we did above with the combined approach.

### Non-uniqueness and Identification

Sometimes the objective does not have a unique maximizer. Consider:
$$ f(x_1,x_2)=-(x_1+x_2-3)^2$$

Any value of $(x_1,x_2)$ that sums to 3 will minimize it. This relates to **identification problems** in estimation.

For example, suppose $x_1$ is *ability* and $x_2$ is *effort*. Together they generate *output* given by $x_1+x_2$. The idea is that $(x^\star_1+x^\star_2)=\text{output}=3$ could be identified in our data. However, whether that output is coming from ability or effort, we cannot discern.

```python
# Non-unique optimizer
def f_saddle(x):
    return (x[0] + x[1] - 3)**2 + (x[0]**2) * 0.000001

result = optimize.minimize(f_saddle, [0, 0], method='BFGS')
print(f"Found: {result.x}")
print(f"Value: {result.fun}")
print(f"Sum of x: {sum(result.x)}")

# Check the Hessian — notice the near-zero eigenvalue
hess = np.linalg.inv(result.hess_inv)
print(f"\nHessian:\n{hess}")
print(f"Eigenvalues: {np.linalg.eigvals(hess)}")
```

```python
# Example: Maximizing g(x) to find the maximum
result_max = optimize.minimize(lambda x: -g(x), 20)  # Negate for maximization

print(f"Maximum of g(x):")
print(f"  x = {result_max.x[0]:.6f}")
print(f"  g(x) = {g(result_max.x[0]):.6f}")
```

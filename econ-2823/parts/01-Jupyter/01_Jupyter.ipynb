## LaTeX

Markdown cells also support LaTeX for writing math. This will be essential for expressing economic models and statistical formulas.

**Inline math** uses single dollar signs: `$x^2$` renders as $x^2$.

**Display math** uses double dollar signs for centered equations:

$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

Some examples you'll see in this course:

- A regression equation: $y_i = \beta_0 + \beta_1 x_i + \varepsilon_i$
- A probability: $P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt$
- Greek letters: $\alpha, \beta, \gamma, \sigma, \mu$

For a full reference, see the [LaTeX math guide](https://en.wikibooks.org/wiki/LaTeX/Mathematics).

## Working with Data

The whole point of this course is to work with real data. We'll use `pandas`, the standard Python library for tabular data, to load a dataset of MLB player statistics.

If you run something without assigning it to a variable, Jupyter will display the output directly.

````python
# Import core data science libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Set up nicer plot defaults
plt.rcParams['figure.figsize'] = (10, 6)
sns.set_style('whitegrid')

# Load the MLB dataset
MLB16_18 = pd.read_csv('https://tayweid.github.io/econ-2823/parts/data/MLB.csv')
````

````python
# View first few rows
MLB16_18.head()
````

````python
# Column names
MLB16_18.columns.tolist()
````

````python
# Data types and basic info
MLB16_18.info()
````

````python
# Summary statistics
MLB16_18.describe()
````

## Linear Regression

In Python, we use `statsmodels` for linear regression. Simply write out the equation. An intercept is included by default.

````python
# Linear regression using statsmodels:

# 1. Define the dataset and the model equation
BA_reg1 = smf.ols('BA ~ Age + Agesqrd', data=MLB16_18).fit() # Use 'BA ~ Age + Agesqrd -1' for no intercept

# 2. Print the model summary table
print(BA_reg1.summary().tables[1])
````

````python
# Full model summary
print(BA_reg1.summary())
````

## Plotting

For plotting, we have two main options:

* **matplotlib** — low-level plotting library with full control
* **seaborn** — higher-level statistical visualization with nicer defaults

````python
# Get unique ages and their fitted values
plot_data = MLB16_18.copy()
plot_data['fitted'] = BA_reg1.fittedvalues

# Sort by Age for line plot
plot_data_sorted = plot_data.groupby('Age')['fitted'].mean().reset_index()

# Plotting fitted values
fig, ax = plt.subplots(figsize=(10, 6))

# Plot line
ax.plot(plot_data_sorted['Age'], plot_data_sorted['fitted'],
        linewidth=2, color='#003594', label='Fitted')

# Plot points
ax.scatter(plot_data_sorted['Age'], plot_data_sorted['fitted'],
           s=50, color='red', zorder=5)

ax.set_xlabel('Age')
ax.set_ylabel('Fitted Batting Average')
ax.set_title('MLB Batting Average: Fitted Quadratic')
plt.tight_layout()
````

````python
# Alternative using seaborn
plt.subplots(figsize=(10, 6))

# Scatter plot with regression line
sns.regplot(x='Age', y='BA', order=2, data=MLB16_18)

# Set Title
plt.title('MLB Batting Average vs Age (with quadratic fit)')
plt.tight_layout()
````

## Notebook Sequentiality

Any cell you run, it's as if you entered that into the Python interpreter. The notebook has a hidden implicit state.

* While the progression of a notebook is meant to be linear from top to bottom, you can run the cells in any order
* You need to make sure you understand which cells you've run and in which order

````python
# Example of sequentiality
y = 4
x = 6
x + y
````

````python
# If you run this after the above, x and y have new values
y = 10
x = 12
x + y
````
